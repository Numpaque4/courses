```python
import numpy as np
import scipy.stats as sc
import matplotlib.pyplot as plt
```

#**Bernoulli**

```python
plt.figure(figsize = (10,6))

p = 0.5
mu, sigma = sc.bernoulli.stats(p, moments = 'mv')

plt.vlines([0,1], [0,0], sc.bernoulli.pmf([0,0], p),
           colors = 'b', lw = 2, alpha = 0.5)

plt.plot([0,1], sc.bernoulli.pmf([0,0], p), 'bo',
         label = f'$\mu_X={mu}\quad\sigma_X^2={sigma}$')

plt.legend()
plt.title(f'Bernoulli\n$p=0.5$')
plt.xlabel('X')
plt.ylabel('$P_X(x)$')
```

#**Geometric**

```python
plt.figure(figsize = (10,6))

p = 0.5
mu, sigma = sc.geom.stats(p, moments='mv')

x = np.arange(sc.geom.ppf(0.01, p),
              sc.geom.ppf(0.99, p))

plt.plot(x, sc.geom.pmf(x, p), 'bo',
         label = f'$\mu_X={mu}\quad\sigma_X^2={sigma}$')

plt.vlines(x, 0, sc.geom.pmf(x, p), colors = 'b', lw = 2, alpha = 0.5)

plt.legend()
plt.title(f'Geometric\n$p=0.5$')
plt.xlabel('X')
plt.ylabel('$P_X(x)$')
```

#**Uniform**

```python
x = np.linspace(0, 10, 10)

plt.figure(figsize = (10,6))

plt.plot(x, [0.1]*10,
         'bo-', lw = 2, alpha= 0.5,
         label = f'$\mu_X={np.mean(x)}\quad\sigma_X^2={0.833}$')

plt.legend()
plt.title(f'Uniform\n$N = 10$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**Binomial Negativa**

```python
n, p = 5, 0.5
mu, sigma= sc.nbinom.stats(n, p, moments='mv')

x = np.arange(sc.nbinom.ppf(0.01, n, p),
              sc.nbinom.ppf(0.99, n, p))

plt.figure(figsize = (10,6))

plt.plot(x, sc.nbinom.pmf(x, n, p), 'bo',
         label = rf'$\mu_X=\bar{{\mu}}_X={mu}\quad\sigma_X^2={sigma}$')

plt.vlines(x, 0, sc.nbinom.pmf(x, n, p), colors = 'b', lw=2, alpha=0.5)

plt.legend()
plt.title(f'Negative Binomial\n$p=0.5\quad r=5$')
plt.xlabel('X')
plt.ylabel('$P_X(x)$')
```

#**HyperGeometric**

```python
[M, n, N] = [20, 7, 12]

mu, sigma = sc.hypergeom.stats(M, n, N, moments = 'mv')

rv = sc.hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_dogs = rv.pmf(x)

plt.figure(figsize = (10,6))

plt.plot(x, pmf_dogs, 'bo',
         label = f'$\mu_X={mu} \quad \sigma_X^2={round(sigma,3)}$')
plt.vlines(x, 0, pmf_dogs, lw=2, alpha = 0.5)

plt.legend()
plt.title(f'Hypergeometric\n$N={M}\quad n={N}\quad N_E={n}$')
plt.xlabel('X')
plt.ylabel('$P_X(x)$')
```

#**Exponencial**

```python
mu, sigma = sc.expon.stats(moments='mv')

plt.figure(figsize = (10,6))

x = np.linspace(sc.expon.ppf(0.01),
                sc.expon.ppf(0.99), 100)

plt.plot(x, sc.expon.pdf(x), 'r-',
         lw = 2, alpha=0.5,
         label=rf'$\mu_X={mu}\quad\bar{{\mu_X}}=\ln 2\quad\sigma_X^2={sigma}$')

plt.legend()
plt.title(f'Exponencial\n$\lambda=1$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**Gamma**

```python
a = 2
mu, sigma = sc.gamma.stats(a, moments='mv')

plt.figure(figsize = (10,6))

x = np.linspace(sc.gamma.ppf(0.01, a),
                sc.gamma.ppf(0.99, a), 100)

plt.plot(x, sc.gamma.pdf(x, a),
       'r-', lw=2, alpha=0.6,
         label=f'$\mu_X={mu}\quad\sigma_X^2={sigma}$')

plt.legend()
plt.title(f'Gamma\n$\\alpha = 2 \quad \\beta=1$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**Chi Square $\chi ^2$**

```python
df = 10
mu, sigma = sc.chi2.stats(df, moments='mv')

plt.figure(figsize = (10,6))

x = np.linspace(sc.chi2.ppf(0.01, df),
                sc.chi2.ppf(0.99, df), 100)
plt.plot(x, sc.chi2.pdf(x, df),
       'r-', lw=2, alpha=0.6,
         label=rf'$\mu_X={mu}\quad\bar{{\mu}}_X=9.33\quad\sigma_X^2={sigma}$')

plt.legend()
plt.title(f'Chi Square $\chi^2$\n$k=10$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**Log - Normal**

```python
s = 1
mu, sigma = sc.lognorm.stats(s, moments='mv')

x = np.linspace(sc.lognorm.ppf(0.01, s),
                sc.lognorm.ppf(0.99, s), 100)

plt.figure(figsize = (10,6))

plt.plot(x, sc.lognorm.pdf(x, s),
       'r-', lw=2, alpha=0.6,
         label=rf'$\mu_X={round(mu,3)}\quad\bar{{\mu}}_X={round(np.exp(mu),3)}\quad\sigma_X^2={round(sigma,3)}$')

plt.legend()
plt.title(f'Log - Normal\n$\sigma=1$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**Beta $\beta$**

```python
a, b = 3, 2
mu, sigma = sc.beta.stats(a, b, moments='mv')

plt.figure(figsize = (10,6))

x = np.linspace(sc.beta.ppf(0.01, a, b),
                sc.beta.ppf(0.99, a, b), 100)
plt.plot(x, sc.beta.pdf(x, a, b),
       'r-', lw=2, alpha=0.6,
         label=f'$\mu_X={round(mu,3)}\quad\sigma_X^2={round(sigma,3)}$')

plt.legend()
plt.title(f'Beta\n$\\alpha=3 \quad \\beta=2$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**t - Student**

```python
df = 3
mu, sigma = sc.t.stats(df, moments='mv')

x = np.linspace(sc.t.ppf(0.01, df),
                sc.t.ppf(0.99, df), 100)

plt.figure(figsize = (10,6))

plt.plot(x, sc.t.pdf(x, df),
       'r-', lw=2, alpha=0.6,
         label=rf'$\mu_X= \bar{{\mu}}_X={round(mu,3)}\quad\sigma_X^2={round(sigma,3)}$')

plt.legend()
plt.title(f't - Student\n$\\nu=3$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```

#**Rayleigh**

```python
mu, sigma = sc.rayleigh.stats(moments='mv')

plt.figure(figsize = (10,6))

x = np.linspace(sc.rayleigh.ppf(0.01),
                sc.rayleigh.ppf(0.99), 100)

plt.plot(x, sc.rayleigh.pdf(x),
       'r-', lw=2, alpha=0.6,
         label=rf'$\mu_X={round(mu,3)}\quad \bar{{\mu}}_X={round(np.sqrt(2*np.log(2)),3)} \quad \sigma_X^2={round(sigma,3)}$')

plt.legend()
plt.title(f'Rayleigh\n$\sigma=1$')
plt.xlabel('X')
plt.ylabel('$f_X(x)$')
```
