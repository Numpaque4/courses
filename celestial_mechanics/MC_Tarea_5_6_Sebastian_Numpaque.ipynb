###**Funcion `lagrange_eqs`**

La siguiente función es construida como método general para encontrar las ecuaciones de movimiento a partir de las transformaciones de coordenadas y las fuerzas involucradas en el sistema.

Para ello se define una variable `system` de tipo `list`, que contiene en cada una de sus entradas a las particulas presentes en el sistema. Cada particula a su vez corresponde a una variable tipo `dict` que contiene el vector de transformación de coordenadas $[x,y,z](\{q_j\})$ y el vector de con las componentes de fuerza sobre la partícula. Los valores de `system` se ingresan como `str` de formato $\LaTeX$ y se emplea la función `parse_latex()` para transformar las entradas `str` a expresiones de `sympy`. El vector de coordenadas en formato `str` debe contener explicitamente la dependencia del tiempo de las variables generalizadas para poder determinar la dependencia temporal de la expresión de `sympy`.

La determinación de la función de *Energía Cinética*, las *Fuerzas Generalizadas* y las ecuaciones de *Lagrange* siguen la teoría provista en la clase.

- **Sobre el sistema de ecuaciones**:

  Las ecuaciones de *Lagrange* tienen la forma
  
  $$0 = f_j (t,\{q_j\}_M,\{\dot q_j\}_M,\{\ddot q_j\}_M)$$

  con la propiedad de ser 'lineales' respecto a $\{\ddot q_j\}_M$. Por esta 'linealidad' es posible construir una forma matricial del sistema de la forma

  $$\mathbf{A\ddot{q}}=\mathbf{b}$$

  donde $\mathbf{A}(t,\{q_j\}_M,\{\dot q_j\}_M)$ es la matriz de coeficientes, $\mathbf{\ddot{q}}$ es el vector de $\{\ddot{q_j}\}_M$ y $\mathbf{b}(t,\{q_j\}_M,\{\dot q_j\}_M)$ es el vector que contiene a los demás terminos lineales asumidos como constantes respecto a $\{\ddot{q_j}\}_M$

  La solución al sistema es directa de modo que cada fila resulte en una solucion del tipo

  $$\mathbf{\ddot{q}}=\mathbf{A^{-1}b}$$

  $$\ddot q_j = f_j (t,\{q_j\}_M,\{\dot q_j\}_M)$$

```python
# @title
def lagrange_eqs(qs, dqs, system):

  #Obtenemos las expresiones de transformacion y Fuerzas del Sistema
  rs = [dic['r'] for dic in system]
  vs = [[x.diff(t) for x in r] for r in rs]
  Fs = [dic['F'] for dic in system]

  #Calculo de Energia Cinetica
  N = len(system)
  masses = np.array(sp.symbols('m_{1:%d}' % (N+1)))
  T = np.sum(0.5*masses*np.array([np.dot(v,v) for v in vs]))

  #Calculo de las Fuerzas Generalizadas
  partial_rs = np.array([[[x.diff(q) for x in r]
                          for r in rs] for q in qs])

  Qs = [np.sum(np.sum(Fs*dr, axis = 1))
                 for dr in partial_rs]

  Qs_simp = np.array([sp.simplify(Q) for Q in Qs])

  #Calculo de las ecuaciones de Lagrange
  partial_Tq = np.array([T.diff(q) for q in qs])
  partial_Tdq = [T.diff(dq) for dq in dqs]
  dpartialT_dt = np.array([dT.diff(t) for dT in partial_Tdq])

  eqs = dpartialT_dt - partial_Tq - Qs_simp
  eqs_simp = [sp.simplify(eq) for eq in eqs]

  #Solucion matricial para d2q
  d2qs = [dq.diff(t) for dq in dqs]

  A = sp.Matrix([[eq.expand().coeff(d2q) for d2q in tuple(d2qs)]
                 for eq in eqs_simp])

  b = A*sp.Matrix(d2qs) - sp.Matrix(eqs_simp)

  d2qs_eqs = (A**-1)*b
  d2qs_eqs_simp = [sp.simplify(d2q_eq) for d2q_eq in d2qs_eqs]

  #Mostramos las ecuaciones
  Nqs = len(qs)
  for i in range(Nqs):
    display(sp.Eq(d2qs[i], d2qs_eqs_simp[i]))
    print('')

  return sp.simplify(T), Qs_simp, d2qs_eqs_simp
```

En estas secciones definimos la variable `system` y realizamos todo el procesamiento para ejecutar la función `lagrange_eqs` con las que obtendremos las ecuaciones de movimiento, la función de energía cinética y las fuerzas generalizadas.

###**Ecuaciones de Lagrange**

```python
# @title
t = sp.Symbol('t')

system = [dict(r = [r'l_1\sin{q_1(t)}',
                    r'-l_1\cos{q_1(t)}',
                    '0'],
               F = [r'T_2\sin{q_2(t)}-T_1\sin{q_1(t)}',
                    r'T_1\cos{q_1(t)}-T_2\cos{q_2(t)}-m_1g',
                    '0']),
          dict(r = [r'l_1\sin{q_1(t)} + l_2\sin{q_2(t)}',
                    r'-l_1\cos{q_1(t)}-l_2\cos{q_2(t)}',
                    '0'],
               F = [r'-T_2\sin{q_2(t)}',
                    r'T_2\cos{q_2(t)}-m_2g',
                    '0'])]

system_latex = [{key: [parse_latex(x) for x in val]
                 for key, val in dic.items()}
                 for dic in system]

Nqs = 2
qs = [dynamicsymbols(f"q_{{{n}}}") for n in range(1, Nqs+1)]
dqs = [q.diff(t) for q in qs]

T, Qs, eqs_L = lagrange_eqs(qs, dqs, system_latex)
```

#**Parte 2 - Lagrangiano**

###**Potencial Generalizado**

**1.** A partir de las fuerzas generalizadas nos proponemos encontrar el potencial generalizado $V$ de modo que podamos escribir estas fuerzas de la forma

$$Q_j=\frac{\partial V}{\partial q_j}$$

Teniendo el siguiente sistema

$$Q_1=gl_1(m_1+m_2)\sin{(q_1)}$$

$$Q_2=gl_2m_2\sin{(q_2)}$$

se cumple que

$$V= \int Q_1\partial q_1=\int Q_1 dq_1 + f(q_2)=-gl_1(m_1+m_2)\cos{(q_1)}+f(q_2)$$

y a su vez se debe satisfacer que

$$\frac{\partial V}{\partial q_2}=Q_2$$

$$\frac{d}{dq_2}f(q_2)=gl_2m_2\sin{(q_2)}$$

por lo que encontramos que

$$f(q_2)=-gl_2m_2\cos{(q_2)}$$

De este modo, la solución al potencial generalizado resulta

$$V=-gl_1(m_1+m_2)\cos{(q_1)}-gl_2m_2\cos{(q_2)}$$

###**Función `euler_lagrange`**

Esta rutina recibe las coordenadas generalizadas, sus derivadas y el *Lagrangiano del sistema* para calcular las ecuaciones de movimiento a partir de él. Iugalmente se calcula la función de *Jacobi*, $h$, dentro de la rutina.

- **Sobre las ecuaciones de movimiento**

  Las ecuaciones de movimiento obtenidas por las ecuaciones de *Euler-Lagrange* tienen la forma

  $$0 = f_j (t,\{q_j\}_M,\{\dot q_j\}_M,\{\ddot q_j\}_M)$$

  por lo que nuevamente proponemos una solución matricial

  $$\mathbf{\ddot{q}}=\mathbf{A^{-1}b}$$

  para obtener ecuaciones de la forma

  $$\ddot q_j = f_j (t,\{q_j\}_M,\{\dot q_j\}_M)$$

```python
# @title
def euler_lagrange(qs, dqs, L):

  #Ecuaciones de Euler Lagrange
  partial_Lq = np.array([L.diff(q) for q in qs])
  partial_Ldq = np.array([L.diff(dq) for dq in dqs])
  dpartialL_dt = np.array([dT.diff(t) for dT in partial_Ldq])

  #Ecuaciones de movimiento
  Nqs = len(qs)
  eqs = dpartialL_dt - partial_Lq
  eqs_simp = [sp.simplify(eq) for eq in eqs]

  #Solucion matricial
  d2qs = [dq.diff(t) for dq in dqs]
  A = sp.Matrix([[eq.expand().coeff(d2q) for d2q in tuple(d2qs)]
                 for eq in eqs_simp])

  b = A*sp.Matrix(d2qs) - sp.Matrix(eqs_simp)
  d2qs_eqs = (A**-1)*b

  #Simplificamos las ecuaciones
  d2qs_eqs_simp = [sp.simplify(d2q_eq) for d2q_eq in d2qs_eqs]

  #Funcion de Jacobi
  h = sp.simplify(np.sum(partial_Ldq*dqs) - L)

  #Mostramos las ecuaciones
  for i in range(Nqs):
    display(sp.Eq(d2qs[i], d2qs_eqs_simp[i]))
    print('')

  return d2qs_eqs_simp, h
```

###**Función `simetries`**

Esta funcion recibe las coordenadas generalizadas, sus derivadas, el lagrangiano y las transformacion local de coordenadas que se desee aplicar. La rutina ejecuta la evaluacion de la variacion del *Lagrangiano* ante la transformación bajo la siguiente teoría

En forma general, se realiza la transformacion de coordenadas

$$q'_j=q_j+\delta K_j(\{q_j\}_M)$$

que son sustituidas al *Lagrangiano* para evaluar su variación respecto a la transformación. El lagrangiano se dice invariante ante la transformación si se satisface que

$$\frac{d}{d \delta}L=0$$

En dicho caso, la cantidad conservada resulta

$$P(\{q_j\}_M)=\sum_j \frac{\partial L}{\partial \dot{q_j}}K_j$$

```python
# @title
def simetries(qs, dqs, Ks, L):

  #Transformacion de Coordenadas
  delta = parse_latex(r'\delta')
  qs_trans = [q + delta*K for q, K in zip(qs, Ks)]
  dqs_trans = [q.diff(t) for q in qs_trans]

  #Sustitucion del Lagrangiano
  L_trans = L.subs(list(zip(qs+dqs,qs_trans+dqs_trans)))

  #Algunas aproximaciones por la variacion local de coordenadas
  cos_approx = [(sp.cos(delta*K), 1) for K in Ks]
  sin_approx = [(sp.sin(delta*K), delta*K) for K in Ks]
  approx = cos_approx + sin_approx + [(delta**2, 0)]

  #Aproximaciones sobre el Lagrangiano
  L_approx = sp.simplify(L_trans.subs(approx))

  #Condicion de Invarianza
  dL = L_approx.diff(delta)

  #Evaluacion de la simetría. Retorna bool
  simetry = dL == 0

  #Calcula la cuadratura de la simetría en lugar de haberla
  quad = (None if simetry == False
          else np.sum([L.diff(dq)*K for dq, K in zip(dqs, Ks)]))

  L_symbol = sp.Function('L')(delta)
  print(f'Hay simetría asociada: {simetry}\n\n')
  display(sp.Eq(L_symbol.diff(delta), dL))
  print('\n\nEsta es su cuadratura:')
  display(quad)

  return quad
```

###**Simetria Traslacional**

Este sistema tiene restricciones y cada particula solamente puede realizar traslaciones que satisfagan

$$x_j^2+y_j^2=l_j^2$$

En otras palabras, la traslación que puede realizar cada particula debe ser puramente angular y lleva la forma

$$q'_j=q_j+\delta q_j$$

$$K_j=q_j$$

Bajo esta transformación, ejecutando la función `simetries`, se encuentra que el *Lagrangiano* no resulta covariante por lo que no se encuentra una cuadratura asociada.

$$\frac{d}{d \delta}L\neq 0$$

```python
# @title

#Desplazamientos
Ks_tras = qs.copy()

#Evaluacion de la simetría
quad_tras = simetries(qs, dqs, Ks_tras, L)
```

###**Simetria Rotacional**

Esta transformación se realiza en un ángulo $\theta$ sobre el plano de movimiento respecto al origen de coordenadas por lo que la transformación se describe de la siguiente forma

$$q'_j=q_j-\delta \theta$$

$$K_j=-\theta$$

El problema anterior se redujo a un a transformación de rotación debido a las reestricciones del sistema donde cada partícula podía trasladarse sobre su trayectoria de forma independiente. En esta transformación de rotación, la rotación es igual e independiente de las coordenadas generalizadas para cada una de las partículas. Con esta rotación, la ejecución de `simetries` nos muestra que el *Lagrangiano* no es covariante por lo que no se encuentra una cuadratura asociada

$$\frac{d}{d \delta}L\neq 0$$

```python
# @title

#Desplazamiento
theta = parse_latex(r'\theta')
Ks_rot = 2*[-theta]

#Evaluacion de la simetria
quad_rot = simetries(qs, dqs, Ks_rot, L)
```

#**Parte 3 - Hamiltoniano**

Para la solución de esta sección, se ha construido la función `hamilton()`, la cual realiza y retorna todos los calculos solicitados

###**Función `hamilton()`**

Esta función recibe las variables de coordenadas y momentos generalizados, juntos con sus derivadas, y las funciones de *Energía Cinética* y *Potencial*.

La función retorna las funciones $\{p_j\}_M(\{q_j\}_M, \{\dot{q_j}\}_M)$, $T_qp$, $L_qp$ y $H$ junto con sus ecuaciones canónicas de movimiento $(\{\dot{p_j}\}_M,\{\dot{q_j}\}_M)$

- **Sobre los Momentos Generalizados**

  Mediante la definición del *Momento Generalizado* se establecen las funciones $\{p_j\}_M(\{q_j\}_M, \{\dot{q_j}\}_M)$

  $$p_j=\frac{\partial L}{\partial \dot{q_j}}$$

  En general, dichas ecuaciones llevan la forma

  $$p_j=f_j(t, \{q_j\}_M, \{\dot{q_j}\}_M)$$

  Al ser un sistema de ecuaciones que es lineal respecto a $\{p_j\}_M$ y $\{\dot{q_j}\}_M$, entonces es posible construir un sistema matricial de la forma

  $$\mathbf{p}=\mathbf{A\dot{q}}$$

  donde $\mathbf{p}$ corresponde al vector de los momentos generalizados, $\mathbf{A}$ equivale a la matriz de coeficientes lineales respecto a $\{\dot{q_j}\}_M$ y $\mathbf{\dot{q}}$ es el vector de las velocidades generalizadas. La solución a este problema resulta inmediata

  $$\mathbf{\dot{q}}=\mathbf{A^{-1}p}$$

  Por tanto existe una transformación a las expresiones de la forma

  $$\dot{q_j}=f_j(t,\{q_j\}_M,\{p_j\}_M)$$

```python
# @title
def hamilton(qs, ps, dqs, dps, T, V):

  #Lagrangiano
  L = T-V

  #Momentos generalizados como funcion de velocidades
  ps_func = [sp.simplify(L.diff(dq)) for dq in dqs]

  #Solución matricial a las velocidades en funcion de los momentos
  M = sp.Matrix([[p.expand().coeff(dq) for dq in tuple(dqs)]
                for p in ps_func])

  dqs_sub = np.array(M**-1) @ ps

  #Sustitucion de las velocidades en T, L, H
  T_qp = sp.simplify(T.subs(list(zip(dqs, dqs_sub))))
  L_qp = sp.simplify(T_qp - V)
  H = sp.simplify(np.sum(ps*dqs_sub) - L_qp)

  #Ecuaciones Canónicas
  dqs_eqs = [sp.simplify(H.diff(p)) for p in ps]
  dps_eqs = [sp.simplify(-H.diff(q)) for q in qs]

  eqs = dqs_eqs + dps_eqs

  #Mostramos las Ecuaciones Canónicas
  for i in range(len(qs)):
    display(sp.Eq(dqs[i], dqs_eqs[i]))
    print('')
    display(sp.Eq(dps[i], dps_eqs[i]))
    print('')

  return ps_func, T_qp, L_qp, H, eqs
```

###**Ecuaciones Canónicas**

Aquí realizamos definición de las variables de momento generalizado $\{p_j\}_M$ junto con sus derivadas y ejecutamos la función `hamilton()`

```python
# @title
#Variables de Momento Generalizado
ps = [dynamicsymbols(f'p_{int(str(q)[3])}') for q in qs]

#Derivadas de Momento Generalizado
dps = [p.diff(t) for p in ps]

#Ejecutamos para obtener funciones
ps_func, T_qp, L_qp, H, eqs_H = hamilton(qs, ps, dqs, dps, T, V)
```

#**Parte 4 - Solución Numérica**

Con la necesidad de dar solución numérica a las ecuaciones de movimiento derivadas del *Lagrangiano* y *Hamiltoniano*, se han definido las funciones `lagrange_solve()` y `hamilton_solve()` que realizan la reducción de orden y linealizacion al sistema de ecuaciones y dan la solución númerica bajo condiciones iniciales.

###**Funcion `lagrange_solve()`**

Esta funcion toma las coordenadas generalizadas y sus derivadas, las ecuaciones de movimiento derivadas del lagrangiano, las condiciones iniciales y el intervalo de integracion. Las ecuaciones de movimiento ya deben traer sustituidas todas las variables constantes en valores numéricos (masas, longitudes y gravedad).

Luego de esto, se realiza la reducción de orden y linealización de las ecuaciones mediante la incorporación de la variable auxiliar $\{u_j\}_M$ velocidad generalizada

$$u_j = \frac{d}{dt}q_j$$

Luego de que el sistema ha sido linealizado, se procede a transformar las expresiones simbólicas en expresiones manipulables mediante la función `sp.lambdify()`. Finalmente son llevadas al integrador de `scipy.integrate.odeint()` y se retornan los valores para $\{q_j\}_M$ y $\{\dot{q_j}\}_M$ solucionados

```python
# @title
def lagrange_solve(qs, dqs, eqs, y0, ts):

  Nqs = len(qs)

  #Variable auxiliar para linealizacion
  us = [dynamicsymbols(f'u_{int(str(q)[3])}') for q in qs]

  #Introducimos la variable auxiliar al sistema de ecuaciones
  dus = [eq.subs(list(zip(dqs, us))) for eq in eqs]

  #Hacemos manipulables las ecuaciones de movimiento
  dqs_func = sp.lambdify(tuple(us), us)
  dus_func = sp.lambdify(tuple(qs+us), dus)

  #Función de integracion
  def ode_system_L(y, ts):

    q = y[:Nqs]
    u = y[Nqs:]

    dq = dqs_func(*u)
    du = dus_func(*y)
    dy = dq + du

    return dy

  #Solucion Numérica
  sol = odeint(ode_system_L, y0, ts)
  qs_vals = sol[:, :Nqs]
  dqs_vals = sol[:, Nqs:]

  return np.array(qs_vals), np.array(dqs_vals)
```

###**Funcion `hamilton_solve()`**

Esta función realiza el mismo proceso que la función `lagrange_solve()`. Se reciben las coordenadas generalizadas, los momentos generalizados, las ecuaciones canónicas, las condiciones iniciales y el intervalo de integracion. Las ecuaciones de movimiento ya deben traer sustituidas todas las variables constantes en valores numéricos (masas, longitudes y gravedad).

A diferencia de la funcion `lagrange_solve()`, esta funcion ya recibe ecuaciones de primer orden por lo que no se debe realizar el procedimiento de reducción de orden ni establecer variables auxiliares.

Directamente se toman las ecuaciones canónicas y se trasnforman a ecuaciones manipulables mediante `sp.lambdify()`. Finalmente son solucionadas por el integrador `scipy.integrate.odeint()` y se retornan los valores para $\{q_j\}_M$ y $\{\p_j\}_M$ solucionados



```python
# @title
def hamilton_solve(qs, ps, eqs, y0, ts):

  Nqs = len(qs)

  #Hacemos manipulables las ecuaciones canónicas
  dqs_func = sp.lambdify(tuple(qs+ps), eqs[:Nqs])
  dps_func = sp.lambdify(tuple(qs+ps), eqs[Nqs:])

  #Funcion de integracion
  def ode_system_H(y, ts):

    dq = dqs_func(*y)
    dp = dps_func(*y)
    dy = dq + dp

    return dy

  #Solucion Numérica
  sol = odeint(ode_system_H, y0, ts)
  qs_vals = sol[:, :Nqs]
  ps_vals = sol[:, Nqs:]

  return np.array(qs_vals), np.array(ps_vals)
```

###**Funcion `system_to_motion()`**

Debido a que el movimiento de las partículas se da en el espacio tridimensional, debemos aplicar las funciones de transformacion de coordenadas expresadas en la variable `system_latex`.

La función `system_to_motion()` toma las constantes dentro de las expresiones y les asigna un valor numérico (masas, longitudes y gravedad); una vez realizado esto, toma las ecuaciones de transformación dentro de `system_latex` y las hace manipulables mediante la función `sp.lambdify()` para los valores de $\{q_j\}_M$ solucionados en el integrador. Con esto retornamos los vectores de posición en el espacio tridimensional para cada valor de $\{q_j\}_M$ solucionado

```python
# @title
def system_to_motion(qs, system, ctes):

  #Tomamos las ecuaciones de transformacion
  rs = [dic['r'] for dic in system]

  #Sustituimos constantes
  rs_subs = [[x.subs(tuple(ctes.items())) for x in r] for r in rs]

  #Ecuaciones manipulables para valores de q_j
  rs_funcs = sp.lambdify(tuple(qs), rs_subs)

  return rs_funcs
```

En esta celda definimos las constantes presentes en las ecuaciones de movimiento y son sustituidas por valores numéricos para ser llevadas a los solucionadores `lagrange_solve()` y `hamilton_solve()`.

Igualmente se ejecuta la función `system_to_motion()` para establecer la función de transformación de coordenadas para describir el movimiento en el espacio tridimensional

```python
# @title
#Constantes de las ecuaciones
ctes = {'m_{1}': 1, 'm_{2}': 1,
        'l_{1}': 1, 'l_{2}': 1,
        'g' : 8}

#Sustituimos constantes en las ecuaciones de movimiento
eqs_L_subs = [eq.subs(tuple(ctes.items())) for eq in eqs_L]
eqs_H_subs = [eq.subs(tuple(ctes.items())) for eq in eqs_H]

rs_func = system_to_motion(qs, system_latex, ctes)
```

En esta celda se establece el intervalo de integracion y las condiciones iniciales, y se da solucion a las ecuaciones de movimiento. Luego, con los valores solucionados de las coordenadas generalizadas, calculamos las posiciones de las partículas en el plano.

```python
# @title
#Intervalo de integracion
Nts = 100
time = 10
ts = np.linspace(0, time, Nts)

#Condiciones iniciales
q0 = np.random.uniform(-np.pi/2, np.pi/2, (2, Nqs))
dq0_p0 = np.random.uniform(-np.pi/4, np.pi/4, (2, Nqs))

y0 = np.hstack((q0, dq0_p0))

#Solucionadores
qs_vals_L, dqs_vals_L = lagrange_solve(qs, dqs, eqs_L_subs, y0[0], ts)
qs_vals_H,  ps_vals_H = hamilton_solve(qs, ps, eqs_H_subs, y0[1], ts)

#Transformamos coordenadas para movimiento en R3
rs_L = np.array([rs_func(*q_val) for q_val in qs_vals_L])
rs_H = np.array([rs_func(*q_val) for q_val in qs_vals_H])

r1s_L = rs_L[:, 0, :]
r2s_L = rs_L[:, 1, :]

r1s_H = rs_H[:, 0, :]
r2s_H = rs_H[:, 1, :]
```

```python
# @title
fig = plt.figure(figsize = (15, 8))
camera = Camera(fig)

ax1 = plt.subplot(121)
ax2 = plt.subplot(222)
ax3 = plt.subplot(224)

ax1.set_xlim(-2.5, 2.5)
ax1.set_ylim(-2.5, 2.5)
ax1.set_box_aspect(1)
ax1.axis('off')

ax2.set_xlim(0, 10)
ax3.set_xlim(0, 10)

ax2.set_title('Lagrange Solution')
ax3.set_title('Hamilton Solution')
ax3.set_xlabel('$t\,(s)$')

for i in range(Nts):

  ax1.plot([0, r1s_L[:,0][i]], [0, r1s_L[:,1][i]], 'k-')
  ax1.plot([r1s_L[:,0][i], r2s_L[:,0][i]],
           [r1s_L[:,1][i], r2s_L[:,1][i]], 'k-')

  ax1.plot(r1s_L[:,0][i], r1s_L[:,1][i], 'bo', markersize=10)
  L_plot, = ax1.plot(r2s_L[:,0][i], r2s_L[:,1][i], 'ro',
           markersize=10)

  ax1.plot([0, r1s_H[:,0][i]], [0, r1s_H[:,1][i]], 'k-')
  ax1.plot([r1s_H[:,0][i], r2s_H[:,0][i]],
          [r1s_H[:,1][i], r2s_H[:,1][i]], 'k-')

  ax1.plot(r1s_H[:,0][i], r1s_H[:,1][i], 'mo', markersize=10)
  H_plot, = ax1.plot(r2s_H[:,0][i], r2s_H[:,1][i], 'go',
          markersize=10)

  ax1.legend([L_plot, H_plot], ['Lagrange', 'Hamilton'],
             loc = 'upper right')


  q1_L, = ax2.plot(ts[:i], qs_vals_L[:,0][:i], 'b-')
  q2_L, = ax2.plot(ts[:i], qs_vals_L[:,1][:i], 'r-')

  ax2.legend([q1_L, q2_L], ['$q_1$', '$q_2$'])

  q1_H, = ax3.plot(ts[:i], qs_vals_H[:,0][:i], 'm-')
  q2_H, = ax3.plot(ts[:i], qs_vals_H[:,1][:i], 'g-')

  ax3.legend([q1_H, q2_H], ['$q_1$', '$q_2$'])

  camera.snap()

plt.close()

animation = camera.animate(interval = time*10)
HTML(animation.to_html5_video())
```

En esta sección sustituimos las variableles constantes en la funcion de *Jacobi* y luego la hacemos manipulable mediante `sp.lambdify()`. A esta función numérica se le ingresan los valores de $\{q_j\}_M$ y $\{\dot{q_j}\}_M$ solucionados en el integrador de `lagrange_solve()` para verificar la invarianza temporal de los valores de la función.

###**Espacio de Fases**

En esta sección realizamos la graficación de los contornos del hamiltoniano para la segunda partícual cuando se ha fijado a la primera partícula en posición $q_1=0$ y $p_1=0$.

El comportamiento de la partícula bajo estas condiciones se reduce al movimiento de un pendulo simple, sin embargo sus contornos de hamiltoniano no adquieren exactamente la forma del espacio de fases del pendulo doble. Esto se debe a que el hamiltoniano ha sido definido en un origen de coordenadas diferente al del punto de anclaje de la partícula y por tanto los contornos cambian.

```python
# @title
#Fijamos valores de q1 y p1
r1_fix = {qs[0]: 0, ps[0]: 0}

#Sustituimos valores constantes en Hamiltoniano
H_subs = H.subs(tuple((r1_fix | ctes).items()))

#Hamiltoniano manipulable
H_func = sp.lambdify((qs[1], ps[1]), H_subs)

#Coordenadas del Espacio de Fase
q2s_phase = np.linspace(-np.pi, np.pi, Nts)
p2s_phase = np.linspace(-5, 5, Nts)

#Contornos
Q2s, P2s = np.meshgrid(q2s_phase, p2s_phase)
H_mesh = H_func(Q2s, P2s)
```

```python
# @title
fig = go.Figure()

fig.update_layout(title = 'Phase Space (Hamiltonian Contour)',
                  xaxis_title = '$q_2$',
                  yaxis_title = '$p_2$',
                  width = 800, height = 600,
                  title_x = 0.5)

fig.add_trace(go.Contour(
              z = H_mesh,
              x = q2s_phase,
              y = p2s_phase,
              colorscale = 'viridis',
              line_smoothing = 0.85,
              contours_coloring='lines',
              line_width=2,
              colorbar=dict(thickness=25,
                            thicknessmode = 'pixels',
                            len = 0.6,
                            lenmode = 'fraction',
                            outlinewidth = 0)))

fig.show()
```

###**Caos del Sistema**

Se ha dicho bastante que este es un sistema sensible a las condiciones iniciales. Al tomar dos sistemas similares que partes de un punto con condiciones iniciales ligeramente diferentes, podemos notar que con el paso del tiempo, los comportamientos de cada sistema se hacen completamente indiferenciables entre sí a pesar de haber tenido un origen común.

Esto se puede apreciar en la evolución de las coordenadas generalizadas para cada sistema donde logramos notar que sus valores cada vez se distancian más entre sí, mostrando que la amplitud de su diferencia crece con el tiempo.

```python
# @title
#Condiciones iniciales
q0 = np.random.uniform(-np.pi/2, np.pi/2, Nqs)
dq0_p0 = np.random.uniform(-np.pi/4, np.pi/4, Nqs)

y0 = np.hstack((np.vstack((q0,q0+0.02)), np.vstack((dq0_p0, dq0_p0+0.02))))

#Solucionador a ambos sistemas
qs_vals_1,  ps_vals_1 = hamilton_solve(qs, ps, eqs_H_subs, y0[0], ts)
qs_vals_2,  ps_vals_2 = hamilton_solve(qs, ps, eqs_H_subs, y0[1], ts)

#Diferencia númerica de las coordenadas
error_qs = qs_vals_1 - qs_vals_2

#Transformacion de coordenadas para movimiento en R3
rs_1 = np.array([rs_func(*q_val) for q_val in qs_vals_1])
rs_2 = np.array([rs_func(*q_val) for q_val in qs_vals_2])

r1s_1 = rs_1[:, 0, :]
r2s_1 = rs_1[:, 1, :]

r1s_2 = rs_2[:, 0, :]
r2s_2 = rs_2[:, 1, :]
```

```python
# @title
fig = plt.figure(figsize = (15, 8))
camera = Camera(fig)

ax1 = plt.subplot(121)
ax2 = plt.subplot(222)
ax3 = plt.subplot(224)

ax1.set_xlim(-2.5, 2.5)
ax1.set_ylim(-2.5, 2.5)
ax1.set_box_aspect(1)
ax1.axis('off')

ax2.set_xlim(0, 10)
ax3.set_xlim(0, 10)

ax2.set_title('Coordinates Evolution')
ax3.set_title('Caos propagation ($\delta q$)')
ax3.set_xlabel('$t\,(s)$')

for i in range(Nts):

  ax1.plot([0, r1s_1[:,0][i]], [0, r1s_1[:,1][i]], 'k-')
  ax1.plot([r1s_1[:,0][i], r2s_1[:,0][i]],
           [r1s_1[:,1][i], r2s_1[:,1][i]], 'k-')

  ax1.plot(r1s_1[:,0][i], r1s_1[:,1][i], 'bo', markersize=10)
  body1_plot, = ax1.plot(r2s_1[:,0][i], r2s_1[:,1][i], 'ro',
           markersize=10)

  ax1.plot([0, r1s_2[:,0][i]], [0, r1s_2[:,1][i]], 'k-')
  ax1.plot([r1s_2[:,0][i], r2s_2[:,0][i]],
          [r1s_2[:,1][i], r2s_2[:,1][i]], 'k-')

  ax1.plot(r1s_2[:,0][i], r1s_2[:,1][i], 'mo', markersize=10)
  body2_plot, = ax1.plot(r2s_2[:,0][i], r2s_2[:,1][i], 'go',
          markersize=10)

  ax1.legend([body1_plot, body2_plot], ['Body A', 'Body B'],
             loc = 'upper right')


  q1_1, = ax2.plot(ts[:i], qs_vals_1[:,0][:i], 'b-')
  q2_1, = ax2.plot(ts[:i], qs_vals_1[:,1][:i], 'r-')
  q1_2, = ax2.plot(ts[:i], qs_vals_2[:,0][:i], 'm-')
  q2_2, = ax2.plot(ts[:i], qs_vals_2[:,1][:i], 'g-')

  ax2.legend([q1_1, q2_1, q1_2, q2_2],
             ['$q_{1A}$', '$q_{2A}$', '$q_{1B}$', '$q_{2B}$'])

  error_q1, = ax3.plot(ts[:i], error_qs[:,0][:i], color = 'orange')
  error_q2, = ax3.plot(ts[:i], error_qs[:,1][:i], color = 'purple')

  ax3.legend([error_q1, error_q2], ['$\delta q_1$', '$\delta q_2$'])

  camera.snap()

plt.close()

animation = camera.animate(interval = time*10)
HTML(animation.to_html5_video())
```
