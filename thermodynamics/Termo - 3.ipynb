#**Caracteristicas de los Ciclos**

---


Los ciclos termodinámicos Otto, Diesel, Brayton y Stirling son utilizados en motores y máquinas térmicas, pero cada uno tiene características y aplicaciones diferentes.

1. **Ciclo Otto:**
   - **Aplicación Principal:** Se utiliza en motores de gasolina utilizados en automóviles.
   - **Principio de Funcionamiento:** Combustión interna con encendido por chispa (bujía).
   - **Relación de Compresión:** Moderada a alta.
   - **Eficiencia Térmica:** Menor en comparación con el ciclo Diesel debido a una relación de compresión más baja.
   - **Tipo de Combustible:** Gasolina.
   - **Etapas Principales:** Admisión, compresión, explosión y escape.

2. **Ciclo Diesel:**
   - **Aplicación Principal:** Se utiliza en motores diésel utilizados en automóviles, camiones y aplicaciones industriales.
   - **Principio de Funcionamiento:** Combustión interna con encendido por compresión.
   - **Relación de Compresión:** Alta (la compresión es el método principal para encender el combustible).
   - **Eficiencia Térmica:** Mayor en comparación con el ciclo Otto debido a la alta relación de compresión.
   - **Tipo de Combustible:** Diésel.
   - **Etapas Principales:** Admisión, compresión, explosión y escape.

3. **Ciclo Brayton:**
   - **Aplicación Principal:** Se utiliza en motores de turbina de gas y plantas de generación de energía.
   - **Principio de Funcionamiento:** Trabajo neto se realiza en una turbina de gas.
   - **Relación de Compresión:** Varía según la aplicación.
   - **Eficiencia Térmica:** Puede ser alta en aplicaciones eficientes.
   - **Tipo de Combustible:** Gas natural, queroseno, gasolina, etc.
   - **Etapas Principales:** Compresión, combustión, expansión en la turbina y escape.

4. **Ciclo Stirling:**
   - **Aplicación Principal:** Motores de Stirling se utilizan en aplicaciones especiales, como generadores de energía térmica.
   - **Principio de Funcionamiento:** Trabajo neto se realiza en un ciclo cerrado con un gas o fluido de trabajo que se expande y contrae.
   - **Relación de Compresión:** Puede variar, pero típicamente baja en comparación con los ciclos Otto y Diesel.
   - **Eficiencia Térmica:** Potencialmente alta debido a la eficiencia en la conversión de calor en trabajo.
   - **Tipo de Combustible:** No utiliza combustión directa, por lo que puede funcionar con una variedad de fuentes de calor.
   - **Etapas Principales:** Compresión, expansión y transferencia de calor en un ciclo cerrado.

#**Eficiencias de los ciclos**

---



(a) Ciclo de Otto:
La eficiencia térmica $\eta$ en función de las temperaturas para el ciclo de Otto es:

$$\eta = 1 - \frac{T_1}{T_2}$$

Donde:
- $T_1$ es la temperatura en el punto de inicio de la compresión (temperatura de admisión).
- $T_2$ es la temperatura en el punto de fin de la compresión (antes de la combustión).

(b) Ciclo de Diesel:
La eficiencia térmica $\eta$ en función de las temperaturas para el ciclo de Diesel es:

$$\eta = 1 - \frac{1}{\gamma}\frac{T_1}{T_2}$$
Donde:
- $T_1$ es la temperatura en el punto de inicio de la compresión (temperatura de admisión).
- $T_2$ es la temperatura en el punto de fin de la compresión (antes de la combustión).
- $\gamma$ es el índice de compresión adiabática.

(c) Ciclo de Joule o de Brayton:
La eficiencia térmica $\eta$ en función de las temperaturas para el ciclo de Joule o ciclo de Brayton es:

$$\eta = 1 - \frac{T_1}{T_2}$$

Donde:
- $T_1$ es la temperatura en el punto de inicio de la compresión (temperatura de entrada del compresor).
- $T_2$ es la temperatura en el punto de fin de la compresión (temperatura antes de la combustión).
- $\gamma$ es el índice de compresión adiabática.

(d) Ciclo de Stirling:
La eficiencia térmica $\eta$ en función de las temperaturas para el ciclo de Stirling es:

$$\eta = 1 - \frac{T_1}{T_2}$$

Donde:
- $T_2$ es la temperatura en la parte caliente del ciclo (donde se realiza la expansión).
- $T_1$ es la temperatura en la parte fría del ciclo (donde se realiza la compresión).

#**Análisis de las expresiones**

---

Como se observa, muchas de estas expresiones comparten estructuras similares. A continuación se establecen algunas condiciones sobre las expresiones.

Debido a que ninguna maquina termica puede alcanzar un 100% de eficiencia, $\eta=1$, se deben restringir los valores de temperatura que se puede tomar para estos calculos.

1. **Ciclo Otto, Ciclo Brayton, Ciclo Stirling:**
    Para lograr que $\eta<1$ se debe satisfacer que $T_1<T_2$, es decir, que la temperatura de ingreso al compresor sea menor que la temperatura de salida del compresor. Su calculo es el mismo pero su diferencia radica en el tipo de combustible que emplea y las temperaturas que alcanzan estos.

    En el *Ciclo Stirling* esto se satisface ya que por definición, $T_1$ corresponde a la temperatura cuando se llega al reservorio de baja temperatura y $T_2$ corresponde a la temperatura cuando se alcanza el reservorio de alta temperatura.

2. **Ciclo Diesel:**
    Similarmente, para lograr que $\eta<1$ se debe satisfacer que $T_1<\gamma T_2$. Para el caso de gas monoatómico, $\gamma = 5/3$, y en el caso de gas diatómico, $\gamma = 7/5$. La razón entre la temperatura de ingreso al compresor y la temperatura de salida del compresor debe ser menor que $\gamma$.

Como observación general, para alcanzar la eficiencia más alta en cada uno de los ciclos, se debe procurar que las diferencias de temperaturas sean altas. Aún así no se logrará alcanzar $\eta=1$ porque para ello se requeriría que la temperatura de salida tienda al infinito o que la temperatura de ingreso sea igual a cero y esto quebranta leyes termodinámicas.

Estos son algunos de los valores típicos de temperatura a los que opera cada ciclo. Con estos valores podemos estar de acuerdo en tomar un valor fijo,$T_1=673.15\,K=400°C$, y variar $T_2$ en aumento para observar el cambio en la eficiencia. Similarmente fijaremos el valor de $T_2=1573.15\,K=1300°C$ y variaremos $T_1$ hacia su disminución.

#**Gas diatómico**

---

Supongamos que todos los ciclos operan con un gas ideal monoatómico ($\gamma \approx 1.67$) y un gas ideal diatómico ($\gamma \approx 1.4$). Consideremos las mismas temperaturas de entrada y salida para cada ciclo.

1. **Ciclo de Otto, Ciclo Diesel y Ciclo Brayton:**
   - Intrinsecamente, aunque las expresiones de eficiencia dependan únicamente de las temperaturas, estas dependen de la relación de compresión del motor (relación entre el volumen máximo y el volumen mínimo)

   $$\frac{T_1}{T_2}=\bigg(\frac{V_2}{V_1}\bigg)^{\gamma - 1}$$

   - Por lo anterior, la eficiencia del ciclo de Otto será más alta para el gas monoatómico, ya que $\gamma$ es mayor que en el caso del gas diatómico

4. **Ciclo de Stirling:**
   - La eficiencia del ciclo de Stirling no depende directamente de $\gamma$, ya que no involucra procesos de compresión y expansión adiabáticos.
   - La eficiencia del ciclo de Stirling puede ser similar para ambos gases ideales, pero la eficiencia real dependerá de otros factores, como las temperaturas relativas y las propiedades específicas del gas utilizado.

En términos generales, para temperaturas dadas, las máquinas térmicas que utilizan un gas ideal monoatómico serán más eficientes que las que utilizan un gas ideal diatómico. Esto se debe a que un gas monoatómico tiene un $\gamma$ más alto, lo que resulta en una mayor eficiencia térmica en los ciclos termodinámicos. Sin embargo, es importante recordar que las condiciones reales pueden variar y que otros factores también pueden influir en la eficiencia de las máquinas térmicas.

En la siguiente tabla observamos los valores tipicos de temperaturas que se tienen en maquinas que operan con estos ciclos sobre gases diatomicos. Similar al proceso anterior, tomamos $T_1=350°C=623.15\,K$ como fijo y variamos $T_2$ en aumento. Luego fijamos $T_2=1000°C=1273.15\,K$ y variamos $T_1$ en disminución.

```python
# @title
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# @title
def eta_otto_bryton_stirling(T1, T2):
    return 1 - T1/T2

def eta_diesel(T1, T2, gamma):
    return 1 - T1/(gamma*T2)
```

```python
# @title
gamma = [5/3, 7/5]
T1m = 673.15
T2m = 1573.15

T1s_m = np.arange(0, T2m, 10)
T2s_m = np.arange(T1m, T2m, 10)

eta_obs_1 = eta_otto_bryton_stirling(T1s_m, T2m)
eta_obs_2 = eta_otto_bryton_stirling(T1m, T2s_m)
eta_d2_m = eta_diesel(T1m, T2s_m, gamma[0])
eta_d1_m = eta_diesel(T1s_m, T2m, gamma[0])

T1_d = 573.15
T2_d = 1273.15

T1s_d = np.arange(0, T2_d, 10)
T2s_d = np.arange(T1_d, T2_d, 10)
eta_d2_d = eta_diesel(T1_d, T2s_d, gamma[1])
eta_d1_d = eta_diesel(T1s_d, T2_d, gamma[1])
```

```python
# @title
plt.figure(figsize = (14,6))

plt.subplot(1, 2, 1)
plt.plot(T1s_m, eta_d1_m, label = "Diesel, $\gamma=5/3$")
plt.plot(T1s_m, eta_obs_1, label = "Otto, Brayton, Stirling")
plt.plot(T1s_d, eta_d1_d, label = "Diesel, $\gamma=7/5$")
plt.title("$\eta$ con $T_2$ fijo")
plt.xlabel("$T_1\,(K)$")
plt.ylabel("$\eta$")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(T2s_m, eta_d2_m, label = "Diesel, $\gamma=5/3$")
plt.plot(T2s_m, eta_obs_2, label = "Otto, Brayton, Stirling")
plt.plot(T2s_d, eta_d2_d, label = "Diesel, $\gamma=7/5$")
plt.title("$\eta$ con $T_1$ fijo")
plt.xlabel("$T_2\,(K)$")
plt.ylabel("$\eta$")
plt.legend()

plt.show()
```

Con estos gráficos comprobamos la imposibilidad de alcanzar una eficiencia $\eta=1$. Esto requiere que $T_1=0\,\lor T_2\to \infty$, lo cual es imposible fisicamente.

En especial, el *Ciclo Diesel* en gases monoatómicos resulta ser el más eficiente cuando se manejan distintas temperaturas frías. Cuando se maneja diversidad de altas temperaturas, no hay diferencia notable entre el *Ciclo Diesel* con cualquier tipo de gas. En lo que respecta a los *Ciclos Otto, Brayton y Stirling* resultan tener la eficiencia más baja.

Las eficiencias más altas se logran cuando las temperaturas en los reservorios de baja de temperatura son bajas. Esto confirma la afirmación de que la eficiencia es mayor cuando las diferencias de temperaturas son mayores; en especial cuando $T_1\to 0$. No se puede garantizar ciegamente si es más sencillo alcanzar temperaturas bajas o temperaturas altas a nivel de industria.
