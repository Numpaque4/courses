```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Datos iniciales
P1 = 1  # Presión en atm
V1 = 13  # Volumen en L
T1 = 300  # Temperatura en K
gamma = 5/3  # Índice adiabático para un gas monoatómico
R = 0.08205746  # Constante de los gases en J/(mol·K)
n = P1*V1/(R*T1) # Cantidad de sustancia en moles (calculada previamente)


# Proceso 2->3 (expansión isotérmica reversible)
P3 = 2  # Presión en atm
V3 = 2 * V1  # Volumen en L
V2 = (P1*(V1**gamma)/(P3*V3))**1.5
P2 = P3*V3/V2
T2 = T1 * (V1/V2)**1.5
T3 = T2 # Temperatura constante en K (isotérmico)
S23 = n*R*np.log(V3/V2)

# Proceso 3->4 (expansión adiabática reversible)
V4 = (P3*(V3**gamma)/(P1*V1))**1.5
P4 = P3*V3/V4
T4 = T3 * (V3/V4)**1.5
S41 = n*R*np.log(V1/V4)

# Crear los puntos del ciclo
V = [V1, V2, V3, V4, V1]
P = [P1, P2, P3, P4, P1]
T = [T1, T2, T3, T4, T1]
#S = [S1, S2, S3, S4, S1]
```

```python
# Crear el diagrama P-V
plt.plot(V,P)
for i in range(4):
  plt.plot(V[i], P[i], "o", label = i + 1)
#plt.title('Diagrama P-V para el Ciclo')
plt.xlabel('Volumen (L)')
plt.ylabel('Presión (atm)')
plt.legend()
plt.grid(True)
plt.savefig("PV.png")
```
