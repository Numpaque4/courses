```
n_H2 = 1000*u.cm**-3
T = 20*u.K
k_B = 0.695/u.cm/u.K
C_10 = 2.6e-11*(u.cm**3/u.s)

n_crit = A_10/C_10

C_01 = C_10*((2*J_10 + 3)/(2*J_10 + 1))*np.exp(-c.h*nu_10/c.k_B/T)

n1_n0= n_H2*C_01/A_10 * (1 + n_H2*C_10/A_10)**-1
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
N = np.array([1e12, 1e13, 1e14, 1e15, 1e16, 1e17, 1e18])

taus = np.array([4.668e-5, 4.668e-4, 4.668e-3,
                 4.668e-2, 4.667e-1, 4.599, 45.43])

T_R = np.array([7.993e-4, 7.991e-3, 7.972e-2,
                7.789e-1, 6.26, 16.3, 16.51])
```

```
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', ['k', 'gray'])  # Blanco y negro
plt.rcParams['grid.linestyle'] = ':'  # Make grid lines dotted
plt.rcParams['grid.linewidth'] = 0.5 # Make grid lines thinner
plt.rcParams['axes.linewidth'] = 0.8 # Reduce axis linewidth


# Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(8, 6))

# Eje inferior: tau vs N (log-log) con N en y
ax1.loglog(taus, N, '--', label=r'$\tau$', color='k')
ax1.set_xlabel(r'$\tau$', fontsize=10)
ax1.set_ylabel(r'$N$ [cm$^{-2}$]', fontsize=10)
ax1.grid(True, which="both", alpha=0.7)


# Eje superior: T_R vs N (log-log) con N en y
ax2 = ax1.twiny()  # Eje x secundario
ax2.loglog(T_R, N, '-.', label=r'$T_R$ [K]', color='gray')
ax2.set_xlabel(r'$T_R\,[\text{K}]$', fontsize=10)

# Add vertical line and label for T_ex
ax2.axvline(x=17, linestyle='--', linewidth=1)
ax2.text(20, N[2], r'$T_{ex}$', verticalalignment='bottom',
         horizontalalignment='right', fontsize=10)


# Create a combined legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='best')

# Añadir título y ajustar layout
plt.tight_layout()

# Guardar la figura en formato adecuado para publicación
plt.show()
```
