#**Ejercicio 1**

```python
def xor(p,q): #Definicion de la operacion XOR
  if p == True:
    output = int(not(q))
  else:
    output = q
  return output
```

```python
def tabla_xor(): #Impresion de la Tabla de Verdad como una secuencia de lineas
  acumula_lineas = ["P     Q   P xor Q"]
  p = [0,0,1,1]
  q = [0,1,0,1]
  for i in range(4):
    linea = str(p[i]) + "     " + str(q[i]) + "      " + str(xor(p[i],q[i]))
    acumula_lineas.append(linea)
  return acumula_lineas
```

```python
lineas = tabla_xor()
for i in range(len(lineas)):
  print(lineas[i])
```

#**Ejercicio 2**

```python
def dec_to_bin(a, n): #Funcion que representa un entero en binario
  lista_bin = []
  if a > ((2**n) - 1): #Condicional por si el número deseado supera su capacidad de expresión en bits
    return f"No es posible expresar dicho número en base Binaria con {n} bits"
  else:
    for i in range(n): #Realiza las transformaciones a binario para n bits
      lista_bin.append(a%2)
      a = a//2
      if a == 1:
        lista_bin.append(1)
        break
    if len(lista_bin) < n: #Si sobran bits de espacio, se llenan con 0
      for i in range(n - len(lista_bin)):
        lista_bin.append(0)
    return lista_bin[-1::-1] #Se invierte el número
```

```python
dec_to_bin(45,8)
```

```python
dec_to_bin(100,8)
```

```python
dec_to_bin(1000,8) #Se comprueba que el numero 1000 supera la capacidad de expresion en 8 bits
```

#**Ejercicio 3**

```python
def float32_to_dec(binario): #Se realiza la conversión de binario con formato IEEE 754 a valor flotante para 32 bits
  signo = int(binario[0]) #Primer bit de signo
  exponente = int(binario[1:9], base = 2) #8 bits de exponente y su transformacion a entero
  fraccionario = 0
  for i in range(23): #23 bits de Mantisa para transformación a flotante
    fraccionario += int(binario[9+i])*(2**(-i-1))
  num = ((-1)**signo)*((1 + fraccionario)*(2**(exponente - 127)))
  return num #Retorna el numero flotante obtenido
```

```python
float32_to_dec("01000000010010010000111111011011") #El numero obtenido es aproximadamente Pi
```

```python
float32_to_dec("00111110101010101010101010101011") #El numero obtenido es aproximadamente 1/3
```

#**Ejercicio 4**

```python
import numpy as np
```

```python
#Se definen los valores en 32 y 64 bits
f32 = np.float32(1/6)
f64 = np.float64(1/6)

suma_32 = 0
suma_64 = 0
for i in range(6): #Se realizan las operaciones
  suma_32 += f32
  suma_64 += f64

#Se calcula error porcentual para mejor visualización del error
error_32 = abs(1-suma_32)*100
error_64 = abs(1-suma_64)*100
f"El error porcentual en 32 bits es {error_32} %", f"El error porcentual en 64 bits es {error_64} %"
```

```python
#Mismo procedimiento, valor y operación distinta
f32 = np.float32(3**(1/3))
f64 = np.float64(3**(1/3))

producto_32 = 1
producto_64 = 1
for i in range(3):
  producto_32 *= f32
  producto_64 *= f64

#Calculo del error porcentual
error_32 = abs(3-producto_32)*100
error_64 = abs(3-producto_64)*100
f"El error porcentual en 32 bits es {error_32} %", f"El error porcentual en 64 bits es {error_64} %"
```
