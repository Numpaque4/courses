#**Bisección**

```python
def biseccion(f,a,b,tol):
  if f(a)*f(b) < 0:
    p = (a+b)/2
    i = 0
    while np.abs(f(p)) > tol:
      p = (a+b)/2
      if f(a)*f(p) < 0:
        b = p
      elif f(b)*f(p) < 0:
        a = p
      i += 1
      print(p, f(p))
    return p, f(p), i
  else:
    return "Error"
```

#**Punto Fijo**

```python
def punto_fijo(f,p,n, tol):
  def g(x):
    return f(x) + x
  def g_prima(x): #Solucion para posibiliad en de Error en Convergencia
    return x-f(x)
  for i in range(n):
    p = g(p)
    if np.abs(f(p)) < tol:
      break
  return p, f(p), i
```

#**Newton - Rapshon**

```python
def newton_raphson(f,df,p,tol):
  i = 0
  while np.abs(f(p)) > tol:
    p = p - (f(p)/df(p))
    i += 1
  return p, f(p), i
```

#**Secante**

```python
def secante(f, p0, p1, resolucion):
  i = 0
  while (np.abs(p1-p0)) > resolucion:
    pn = p1 - ((f(p1)*(p1-p0))/(f(p1)-f(p0)))
    p0 = p1
    p1 = pn
    i += 1
    print(pn, f(pn), i)
  return pn, f(pn), i
```

#**Punto Falso**

```python
def punto_falso(f, a, b, resolucion):
  if f(a)*f(b) >= 0:
    return 'Error'
  n=0
  while np.abs(a - b) > resolucion:
    p = b - (f(b)*(b-a))/(f(b) - f(a))
    if f(a)*f(p)<0:
      b = p
    if f(p)*f(b)<0:
      a = p
    n += 1
  return p, f(p), n
```

#**Interpolacion Lineal**

```python
def interpolacion_lineal(x, xn, yn):
  y = np.zeros(len(x))
  for j in range(len(x)):
    for i in range(len(xn) - 1):
      if xn[i] <= x[j] <= xn[i+1]:
        m = (yn[i+1] - yn[i])/(xn[i+1] - xn[i])
        y[j] = (m * (x[j] - xn[i])) + yn[i]
        break
  return y
```

#**Interpolación Lagrange**

```python
def interpolacion_lagrange(x, xn, yn):
  n = len(xn) - 1

  def L(x, n, i):
    productoria = 1
    for k in range(n):
      if k != i:
        productoria *= (x - xn[k])/(xn[i] - xn[k])
    return productoria

  sumatoria = 0
  for i in range(n):
    sumatoria += yn[i] * L(x,n,i)
  return sumatoria
```

```python
def polinomio_lagrange(xn, yn):
  n = len(xn)
  x = sp.symbols("x")
  def L(x, n, i):
    productoria = 1
    for k in range(n):
      if k != i:
        productoria *= (x - xn[k])/(xn[i] - xn[k])
    return productoria

  sumatoria = 0
  for i in range(n):
    sumatoria += yn[i] * L(x, n,i)
  return sp.expand(sumatoria, x)
```

#**Diferencias Divididas**

```python
def interpolacion_diferencias(x,xn,yn):
  N = len(xn)
  D = np.zeros((N,N))
  D[0,:] = yn

  #Calculo de Diferencias
  for k in range(1,N):
    for i in range(N-k):
      D[k,i] = (D[k-1,i+1] - D[k-1,i])/(xn[i+k] - xn[i])

  #Aplicación de Logaritmo
  sumatoria = D[0,0]
  for k in range(1,N):
    productoria = 1
    for p in range(k):
      productoria *= (x - xn[p])
    sumatoria += D[k,0] * productoria
  return sumatoria
```

#**Hermite**

```python
def interpolacion_hermite_primera(x,xn,yn,dyn):

  def duplica_entradas(a):
    b = np.zeros(2*len(a))
    for i in range(len(a)):
      b[2*i], b[(2*i) + 1] = a[i], a[i]
    return b

  duplica_xn = duplica_entradas(xn)
  duplica_yn = duplica_entradas(yn)
  duplica_dyn = duplica_entradas(dyn)
  D = np.zeros((len(duplica_xn),len(duplica_xn)))
  D[0,:] = duplica_yn

  #Calculo de Diferencias
  for i in range(len(duplica_yn) - 1):
    if i%2 == 0:
      D[1:i] = duplica_yn[i]
    else:
      D[1,i] = (duplica_yn[i+1] - duplica_yn[i])/(duplica_xn[i+1] - duplica_xn[i])

  for k in range(2,len(duplica_yn) - 1):
    for i in range(len(duplica_yn) - k):
      D[k,i] = (D[k-1,i+1] - D[k-1,i])/(duplica_xn[i+k] - duplica_xn[i])

  #Aplicación de Logaritmo
  sumatoria = D[0,0]
  for i in range(1, len(duplica_yn)):
    productoria = 1
    for j in range(i):
      productoria *= (x - duplica_xn[j])
    sumatoria += D[i,0] * productoria
  return sumatoria
```

#**Splines Cubicos**

```python
def splines_cubicos_free(xn,yn):
  x = sp.symbols("x")
  S = []

  #Primera condición
  for j in range(len(xn) - 1):
    string="a%1.f b%1.f c%1.f d%1.f" % (j,j,j,j)
    a,b,c,d = sp.symbols(string)
    Sj = a + (b*(x - xn[j])) + (c*((x - xn[j])**2)) + (d*((x - xn[j])**3))
    valor_a = sp.solve(Sj.subs(x, xn[j]) - yn[j])[0]
    Sj = Sj.subs(a, valor_a)
    valor_b = sp.solve(Sj.subs(x, xn[j+1]) - yn[j+1],b)[0]
    Sj = Sj.subs(b,valor_b)
    S.append(Sj)

  #Segunda condicion
  for j in range(len(xn) - 2):
    string="a%1.f b%1.f c%1.f d%1.f" % (j,j,j,j)
    a,b,c,d = sp.symbols(string)
    dSj = sp.diff(S[j], x)
    dSj_1 = sp.diff(S[j+1], x)
    delta_dS = (dSj - dSj_1).subs(x, xn[j+1])
    valor_c = sp.solve(delta_dS, c)[0]
    S[j] = S[j].subs(c, valor_c)
    for i in range(0,j):
      S[i]=S[i].subs(c,valor_c)

  for j in range(len(xn) - 2):
    string="a%1.f b%1.f c%1.f d%1.f" % (j,j,j,j)
    a,b,c,d = sp.symbols(string)
    d2Sj = sp.diff(S[j], x, 2)
    d2Sj_1 = sp.diff(S[j+1], x, 2)
    delta_d2S = (d2Sj - d2Sj_1).subs(x, xn[j+1])
    valor_d = sp.solve(delta_d2S, d)[0]
    S[j] = S[j].subs(d, valor_d)
    for i in range(0,j):
      S[i]=S[i].subs(d,valor_d)

  #Fronteras
  string = "c%1.f d%1.f" % (len(xn) - 2, len(xn) - 2)
  c, d = sp.symbols(string)
  d2Sj = sp.diff(S[0], x, 2)
  valor_c = sp.solve((d2Sj.subs(x, xn[0])), c)[0]
  for j in range(len(xn) - 1):
    S[j] = S[j].subs(c, valor_c)
  d2Sj_1 = sp.diff(S[-1], x, 2)
  valor_d = sp.solve((d2Sj_1.subs(x, xn[-1])), d)[0]
  for j in range(len(xn) - 1):
    S[j] = S[j].subs(d, valor_d)
  for j in range(len(xn) - 1):
    S[j] = S[j].expand()
  return S
```

#**Adelante**

```python
def dif_adelante(f,t,x,tol):
  h = np.zeros(10)
  for i in range(0,10):
    h[i] = 10**(-i+1)
  df = sp.lambdify(t, sp.diff(f,t), 'math') #También usar Numpy para que sea sobre un arreglo
  for i in range(len(h)):
    lim = (f(x+h[i]) - f(x))/h[i]
    if np.abs(lim - df(x)) < tol:
      return lim
```

#**Atrás**

```python
def dif_atras(f,t,x,tol):
  h = np.zeros(10)
  for i in range(0,10):
    h[i] = 10**(-i+1)
  df = sp.lambdify(t, sp.diff(f,t), 'math') #También usar Numpy para que sea sobre un arreglo
  for i in range(len(h)):
    lim = (f(x) - f(x - h[i]))/h[i]
    if np.abs(lim - df(x)) < tol:
      return lim
```

#**Central**

```python
def dif_central(f,t,x,tol):
  h = np.zeros(10)
  for i in range(0,10):
    h[i] = 10**(-i+1)
  df = sp.lambdify(t, sp.diff(f,t), 'math') #También usar Numpy para que sea sobre un arreglo
  for i in range(len(h)):
    lim = (f(x+h[i]) - f(x-h[i]))/(2*h[i])
    if np.abs(lim - df(x)) < tol:
      return lim
```

#**Riemann**

```python
def integral_riemann(xn, yn):
  integral = np.zeros(len(xn) - 1)
  h = (xn[-1] - xn[1])/len(xn)
  for i in range(len(integral)):
    integral[i] = yn * h
  return np.sum(integral)
```

#**Trapecio**

```python
def integral_trapecio(xn, yn):
  integral = np.zeros(len(xn) - 1)
  for i in range(len(integral)):
    integral[i] = ((yn[i] + yn[i+1]) * (xn[i+1] - xn[i]))/2
  return np.sum(integral)
```

#**Simpson $\frac{1}{3}$**

```python
def integral_simpson(xn, yn):
  integral = np.zeros((len(xn) - 1)/2)
  for i in range(len(integral)):
    h = (xn[(2*i)+2] - xn[2*i])/2
    integral[i] = (h/3) * (yn[2*i] + (4*yn[(2*i)+1]) + yn[(2*i)+2])
  return np.sum(integral)
```

#**Operaciones Elementales**

```python
def intercambio_filas(A,i,j):
  B = (np.copy(A)).astype("float64")
  B[[i,j]] = B[[j,i]]
  return B

def escalar_filas(A,i,lambda_):
  B = (np.copy(A)).astype("float64")
  B[i,:] = lambda_*B[i,:]
  return B

def suma_filas(A,i,j,lambda_):
  B = (np.copy(A)).astype("float64")
  B[i,:] += lambda_ * B[j,:]
  return B
```

#**Eliminación Gaussiana**

```python
def eliminacion_gaussiana(A):
  B = (np.copy(A)).astype("float64")
  m = len(B[:,0])
  n = len(B[0,:])

  for j in range(min(m,n)):
    for i in range(j,m): #Coloca Pivotes
      if B[i,j] != 0:
        B = intercambio_filas(B,i,j)
        break
    for i in range(j+1, m): #Reduce Matriz a Triangular Inferior
      if B[i,j] != 0:
        B =  suma_filas(B, i, j, -B[i,j]/B[j,j])
  for i in range(min(m,n)):
    if B[i,i] != 0:
      B = escalar_filas(B,i,1/B[i,i])
  return B
```

#**Solución Gausiana**

```python
def solucion_gauss(A,b):
  n = len(A[0,:])
  B = eliminacion_gaussiana(np.c_[A,b])
  A_prima = B[:,0:n]
  b_prima = B[:,-1]
  x = np.zeros(len(A[:,0]))

  for i in range(n-1, 0-1, -1):
    if A_prima[i,i] == 0:
      x[i] = 2 #Matriz LD (Se puede modificar)
    else:
      x[i] = (b_prima[i] - sum(A_prima[i,i+1:n]*x[i+1:n]))/(A_prima[i,i])

  return x
```

#**Gauss - Jordan**

```python
def gauss_jordan(A):
  B = eliminacion_gaussiana(A)
  m = len(B[:,0])
  n = len(B[0,:])

  for j in range(min(m,n)-1,0,-1):
    for i in range(j-1, -1, -1): #Reduce Matriz a Triangular Inferior
      if B[i,j] != 0:
        B =  suma_filas(B, i, j, -B[i,j]/B[j,j])

  return B
```

#**Matriz Inversa**

```python
def inversa(A):
  I = np.identity(len(A))
  A_aumentada = np.c_[A,I]
  solucion = gauss_jordan(A_aumentada)
  A_inversa = solucion[:, len(A):]
```

#**Determinante**

```python
def determinante(A):
  B = (np.copy(A)).astype('float64')
  n = len(B)
  if n == 1:
    return B[0,0]
  det = 0
  elimina_fila = np.delete(np.arange(n),0)
  for j in range(n):
    elimina_columna = np.delete(np.arange(n),j)
    menor = B[elimina_fila,:][:,elimina_columna]
    cofactor = B[0,j]*((-1)**j)
    det += cofactor * determinante(menor)
  return det
```

#**Método de Euler**

```python
def euler(xn, y0, h):
  yn = np.zeros()
  yn[0] = y0
  for i in range(len(xn)):
    yn[i+1] = y0 + h*f(xn[i], yn[i])
  return yn
```

##**Runge Kutta - Segundo Orden**

```python
def RK2(f, y, t, dt):
  k1 = dt*f(y,t)
  k2 = dt*f(y + 0.5*k1, t + 0.5*dt)
  return y + k2
```

#**Runge Kutta - Cuarto Orden**

```python
def RK4(f, y, t, dt):
  k1 = dt*f(y,t)
  k2 = dt*f(y + 0.5*k1, t + 0.5*dt)
  k3 = dt*f(y + 0.5*k2, t + 0.5*dt)
  k4 = dt*f(y + k3, t + dt)
  return y + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
```
