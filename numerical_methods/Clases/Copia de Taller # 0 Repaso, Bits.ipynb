# Funciones y Gráficas

````python
#Definición explicita
def f(x):
    return np.exp(-x)*np.cos(x)*np.sin(x)

x=np.arange(0,2*np.pi,0.1)
y=f(x)

plt.figure(figsize=(8,6)) #largo, alto
plt.title('Título de la figura: $f(x)$')
plt.xlabel('$x (m)$')
plt.ylabel('$y (a.u)$')

plt.plot(x,y,'o--',color='blue',label='$e^{-x} \cos{(x)} \sin{(x)}$')
plt.plot(x,np.exp(-x)*x**2,color='red',label='$e^{-x} x^2$')
plt.legend()
plt.grid()
````

# Unidades básicas de computación e información

La unidad básica de computación e información es el $\textbf{bit}$ y puede tomar valores binarios, estos valores binarios se les asigna $\textbf{0,1}$

Un número $\textbf{entero}$ puede descomponerse en su representación binaria:

```c
       10/2     0
       5/2      1 
       2/2      0
       1/2      1
```  

$10 = 1\times2^3 + 0\times2^2 + 1\times2^1 + 0\times2^0$

$10=1010_2$

$10.3125 = 1010..._2$

Un número de $\textbf{punto flotante}$ también puede descomponerse en su representación binaria:

```c
        Parte entera          Parte fracional
       10/2     0          0.3125*2=0.625     0     
       5/2      1           0.625*2=1.25      1
       2/2      0            0.25*2=0.5       0
       1/2      1             0.5*2=1.0       1
```  

$10 = 1\times2^3 + 0\times2^2 + 1\times2^1 + 0\times2^0$

$0.3125 = 0\times2^{-1} + 1\times2^{-2} + 0\times2^{-3} + 1\times2^{-4}$

$10.3125 = 1010.0101_2$

````python
a=int(10) #parte entera
b=0.3125 #parte fraccional

print('Numpy:',np.binary_repr(10)) #funcion de numpy que lo hace automatico para numeros ENTEROS

print('Parte entera')
cont=0
while a>=1:
  print(a,a%2)
  a=a//2
  cont+=1
  
  
print('---------------')

print('Parte fracional')

while b<1:
  print(b,b*2,int(b*2))
  b=b*2
  if b>1:
    b=b-1

print('-------------------')

print('Numero representado')
print(1*2**3 + 0*2**2 + 1*2**1 + 0*2**0 + 0*2**-1 + 1*2**-2 + 0*2**-3 + 1*2**-4)
print('1010.0101')
````

###### 32 bits

$$(-1)^s (1+f)2^{e-127}$$

$$s=b_{31}$$

$$f=\sum_{i=1}^{23} b_{23-i} 2^{-i}$$

$$e=\sum_{i=0}^7 b_{23+i} 2^{i}$$

````python
print('mínimo exponente',0)
print('máximo exponente',2**(8)-1)
print('mínimo número',np.float32(2**-127))
````

````python
def bits32(s,f,e):
  return (((-1)**s)*((1+f))*(2.**(e-127)))
````

````python
#Número random
b=np.random.choice([0,1],32)
print(b[::-1])

s=b[31]

f=0
for i in range(1,23+1):
  f+=b[23-i]*2**-i

e=0
for i in range(0,7+1):
  e+= b[23+i]*2**i

print(s,f,e)
print(bits32(s,f,e))
````

````python
#Máximo número
b=np.zeros(32)+1
print(b[::-1])

s=b[31]

f=0
for i in range(1,23+1):
  f+=b[23-i]*2**-i

e=0
for i in range(0,7+1):
  e+= b[23+i]*2**i

print(s,f,e)
print(bits32(s,f,e))
````

###### 64 bits

$$(-1)^s (1+f)2^{e-1023}$$

$$s=b_{63}$$

$$f=\sum_{i=1}^{52} b_{52-i} 2^{-i}$$

$$e=\sum_{i=0}^{10} b_{52+i} 2^{i}$$

````python
#Minimos
print('32 bits', np.float32(2**-127))
print('64 bits', 2**-1023)
````

$$\frac{1}{9}+\frac{1}{9}+\frac{1}{9}+\frac{1}{9}+\frac{1}{9}+\frac{1}{9}+\frac{1}{9}+\frac{1}{9}+\frac{1}{9}=1$$

````python
uno=0
for i in range(9):
  uno+=1/9

print(1/9)
print(uno)
````

$$2^{1/5}\times2^{1/5}\times2^{1/5}\times2^{1/5}\times2^{1/5}=2$$

````python
dos=1
for i in range(5):
  dos*=2**(1/5)

print(dos)
````
