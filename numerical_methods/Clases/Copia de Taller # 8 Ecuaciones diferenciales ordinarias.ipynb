# Sistema de ecuaciones diferenciales ordinarias

Un sistema de ecuaciones diferenciales es un conjunto de ecuaciones diferenciales de primer orden acopladas en las funciones $x_0,x_1,...x_n$ dependientes del tiempo $t$:

$$\frac{d x_0}{dt} = g_0(x_0,x_1,...,t) \\
\frac{d x_1}{dt} = g_1(x_0,x_1,...,t) \\
\vdots \\
\frac{d x_n}{dt} = g_n(x_0,x_1,...,t)$$

### Recordemos derivadas numéricas

##### Notación indicial
$t_i = t_0+i\Delta t $

$f_i=f(t_i)$

##### Diferencia hacia adelante (Forward difference)
$f'_i=\frac{f_{i+1}-f_i}{\Delta t}$

##### Diferencia central (Central difference)
$f'_i=\frac{f_{i+1}-f_{i-1}}{2\Delta t}$

#### Segunda derivada

$f''_i=\frac{f_{i+1}+f_{i-1}-2f_i}{{\Delta t}^2}$

# Sistema físico: Oscilador armónico amortiguado

El sistema físico de estudio será el oscilador armónico amortiguado, que está descrito por la ecuación diferencial de segundo orden:

$$\frac{d^2 x}{dt^2} + \gamma \frac{d x}{dt} + \omega^2 x = 0$$

donde $\omega$ es la frecuencia natural del sistema y $\gamma$ el coeficiente de fricción.

Haciendo el cambio de variable $\frac{dx}{dt}=v$, podemos cambiar la ecuación diferencial de segundo orden, por dos de primer orden:

$$\frac{dx}{dt}=v = g_0(x,v,t)$$
$$\frac{dv}{dt}=-\gamma v - \omega^2 x=g_1(x,v,t)$$


```python
#Parámetros físicos (En un Sistema de Unidades conveniente a las escalas del sistema)

gamma=1 #Coeficiente de friccion
omega=1 #Frecuencia natural

def g0(x,v,t):
    return v
def g1(x,v,t):
    return -gamma*v-omega**2*x
```

# Método de Euler
Consiste en cambiar los diferenciales por diferencias hacia adelante de la forma:
$f'_i=\frac{f_{i+1}-f_i}{\Delta t}$

Las ecuaciones del oscilador armónico se convierten en:

$$\begin{align}\frac{dx(t)}{dt}&=v(t)  &\to \frac{x_{i+1}-x_i}{\Delta t}&=v_i=g_0(x_i,v_i,t_i)\\
\frac{dv(t)}{dt}&=-\gamma v(t) - \omega^2 x(t) &\to \frac{v_{i+1}-v_i}{\Delta t}&= -\gamma v_i - \omega^2 x_i = g_1(x_i,v_i,t_i)
\end{align}$$

Despejando los tiempos futuros $i+1$ se obtiene:

$$x_{i+1}=g_0(x_i,v_i,t_i)\Delta t + x_i$$
$$v_{i+1}=g_1(x_i,v_i,t_i)\Delta t+v_i$$

#### Sin amortiguamiento

$$\gamma=0$$

#### Sobreamortiguado

$$ \gamma^2 -4 m^2 \omega^2>0$$

#### Amortiguamiento crítico

$$ \gamma^2 -4 m^2 \omega^2=0$$

#### Amortiguado

$$ \gamma^2 -4 m^2 \omega^2<0$$


```python
omega=1 #Frecuencia natural
m=1 #masa
gammas=[2*m*omega+1.5,2*m*omega,2*m*omega-1.5,0] #Coeficientes de friccion

for gamma in gammas:

    dt=0.1
    t0=0
    tf=20

    t=np.arange(t0,tf+dt,dt)
    x=np.zeros(len(t))
    v=np.zeros(len(t))

    x[0]=2
    v[0]=0

    y=[x[0],v[0]]

    for i in range(len(t)-1):
        y=RK4(function,y,t[i],dt)
        x[i+1]=y[0]
        v[i+1]=y[1]

    if gamma==gammas[0]:
        label='Sobreamortiguado'
    elif gamma==gammas[1]:
        label='Amortiguamiento crítico'
    elif gamma==gammas[2]:
        label='Amortiguado'
    elif gamma==gammas[3]:
        label='No amortiguado'

    plt.plot(t,x,label=label)
    plt.xlabel('$t$')
    plt.ylabel('$x(t)$')

plt.hlines(0,t0,tf,linestyles='dashed')
plt.legend()
```
