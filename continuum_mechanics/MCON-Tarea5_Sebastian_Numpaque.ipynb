5. **Loopback time**. Hay una forma de calcular la edad del
   universo, que en los experimentos anteriores tuvo que encontrarse manualmente.  Para ello se define el que se conoce como el *loopback time* (tiempo de mirada atrás), $t_{LB}$. Este número es el tiempo que se ha expandido el universo desde un valor específico del factor de escala $a(t)$ hasta el presente cuando $a(t_0)=1$.

   Partiendo de la definición del parámetro de Hubble:
   $$
   H=\frac{\dot a}{a}
   $$

   **mostrar** que el tiempo cósmico correspondiente a un determinado valor de $a$ se puede obtener como:

   $$
   t_{LB}(a) = t_0 - t(a) = \int_{a}^{1} \frac{\mathrm{d}a'}{a' H(a')}
   $$
   donde $t_0$ es el tiempo actual. Nótese que, por definición, cuando $a=0$ el $t_{LB}$ es la edad $\tau_0$ del universo.

6. **Evolución de H**. Para calcular el $t_{LB}$ es necesario encontrar
   una ecuación que nos de $H$ como función de $a$. Eso se logra despejando $H$ de la ecuación de Friedman-Lemeitre:

   $$
   \epsilon=\frac{4 \pi}{3} G a^2\left[\frac{3 H^2}{8 \pi G}-\left(\rho+\rho_{\Lambda}\right)\right]
   $$
   donde recordemos $\epsilon$ es una constante.

   **Demuestre** que $H$ se puede escribir como:

   $$
   \frac{H^2}{H_0^2} = \frac{\rho+\rho_\Lambda + \rho_k}{\rho_{c0}}
   $$

   donde:
   $$
   \rho_k \equiv \frac{3 \epsilon}{4\pi G} \frac{1}{a^2}
   $$
   es una densidad ficticia que llamaremos *densidad de curvatura* (por un término usado en relatividad general, pero que aquí es solo una definición).

   Sabemos que la materia tiene una densidad que cambia según la ley $\rho_m=\rho_{0m}(a_0/a)^3$, que la energía oscura tiene $\rho_\Lambda = \rho_{0\Lambda}$ (no cambia en el tiempo) y que la densidad de curvatura varia según $\rho_k=\rho_{0k}(a_0/a)^2$. Usando esto mostrar que $H$ se puede escribir como:

   $$
   H(a) = H_0\sqrt{\Omega_{0m}(a_0/a)^3 + \Omega_{0k}(a_0/a)^2 + \Omega_\Lambda}
   $$

7. **Edad del universo teórica**. En astrofísica, el corrimiento al rojo
   se relaciona con el factor de escala así: $a/a_0 = 1/(1+z)$.  
   
   **Demuestre**, teniendo en cuenta la ecuación para $H(a)$ y escribiéndola en términos de $z$ el $t_{LB}$ es:

   $$
   t_{LB}(z) = t_H \int_0^z \frac{dz}{(1+z)\sqrt{\Omega_{0m}(1+z)^3 + \Omega_{0k}(1+z)^2 + \Omega_\Lambda}}
   $$

   De allí, la edad del universo será finalmente:

   $$
   \tau_0 = t_H \int_0^\infty \frac{dz}{(1+z)\sqrt{\Omega_{0m}(1+z)^3 + \Omega_{0k}(1+z)^2 + \Omega_\Lambda}}
   $$

   Donde $t_H=1/H_0$.

#**Exercise 1**

Desde el análisis dimensional, las unidades de las siguientes constantes vienen dadas sobre el Sistema Internacional

$$[G]=\frac{[m^3]}{[kg][s^2]}\quad[c]=\frac{[m]}{[s]}\quad[H_0]=\frac{1}{[s]}$$

Si definimos su valor como $G=H_0=c=1$ formamos un sistema de ecuaciones con el que determinamos la unidad de masa, longitud, tiempo y densidad del sistema.

$$\mathcal{U}_T =\frac{1}{H_0}\quad \mathcal{U}_L=\frac{c}{H_0}\quad\mathcal{U}_M=\frac{c^3}{G\,H_0}\quad\mathcal{U}_\rho = \frac{H_0^2}{G}$$

Empleamos el módulo `astropy.units` para realizar el manejo de unidades de forma más simple.

```python
# @markdown Canonical Units

#S.I. Values
c = cons.c
G = cons.G
H0 = 70*u.km/u.s/u.Mpc

#Defining Units
Ut = u.def_unit('Ut', (1/H0).to(u.s))
UL = u.def_unit('UL', (c/H0).to(u.m))
UM = u.def_unit('UM', (c**3/(G*H0)).to(u.kg))
Urho = u.def_unit('Urho', (H0**2/G).to(u.kg/u.m**3))

#Reset Values
c, G, H0 = (1,1,1)

print('Las Unidades Canónicas del Sistema son:\n\n' \
      f'Ut = {Ut.decompose()}\nUL = {UL.decompose()}\n'\
      f'UM = {UM.decompose()}\nUρ = {Urho.decompose()}')
```

#**Exercise 2**

Definimos la función `fluid_universe` como función de integración del sistema de ecuaciones

$$
  \begin{aligned}
\dot{\rho} & =-3 \rho H(t) \\
\dot{H}+H^2 & =-\frac{4 \pi}{3} G \rho(t)
\end{aligned}
  $$

```python
# @markdown Cosmological Model

def fluid_universe(t, y):

  rho, H = y

  drho = -3*rho*H
  dH = -4*np.pi*G*rho/3 - H**2

  return np.array([drho, dH])
```

Realizamos la integración hacia el futuro hasta $t_{[Ut]}=1$, es decir, integrando hasta el tiempo de Hubble hacia el futuro.

Todo el calculo se realiza bajo Unidades Canónicas y los valores de Densidad y Parámetro de Hubble se almacenan en dichas unidades.

```python
# @markdown Fluid Universe Equations Solution

#Initial Conditions
rho_0 = (3e-27*u.kg/u.m**3).to(Urho).value #Uρ

#Equations Solution
universe_future = solve_ivp(fluid_universe, (0, 1), [rho_0, H0],
                            max_step = 0.01)

#Solution Values
ts_future = universe_future.t*Ut.to(u.Gyr) #Gyr
rhos_future, Hs_future = universe_future.y
```

A continuación retornamos al sistema de unidades convencional para mostrar más apropiadamente los valores como suelen encontrarse dentro de la literatura.

Bajo esta solución de la consideración del Universo como Medio Continuo, el Universo tiende a enfriarse y las Galaxias se alejan a una tasa cada vez menor.

 **Nota:** El ploteo de la solución en unidades convencionales es local. No se cambian las unidades de las soluciones en sus variables respectivas

```python
# @markdown Plotting Future Parameters

go.Figure(data = [go.Scatter(x = ts_future,
                             y = rhos_future*Urho.to(u.kg/u.m**3)*1e28,
                             name = 'ρ(t)'),
                  go.Scatter(x = ts_future,
                             y = Hs_future/Ut.to(u.s*u.Mpc/u.km),
                             name = 'H(t)')],

          layout = go.Layout(width = 900, height = 600,
                             title_x = 0.5, title_y = 0.95,
                             title_text = "Density and Hubble's " \
                                          "Parameter",
                             xaxis_title = 't [Gyr]',
                             yaxis_title = 'ρ [1e-28 kg/m^3],   ' \
                                           'H [km/s/Mpc]',
                             legend = dict(orientation = 'h',
                                           y = 1.08, x = 0.8)))
```

#**Exercise 3**

El método **LSODA** se enfoca principalmente en la solución de **Ecuaciones Stiff**, haciendo referencia a Ecuaciones Diferenciales númericamente inestables y que requieren de un tamaño de paso de integración extremadamente pequeño.

Por la naturaleza de las ecuaciones y su divergencia en la densidad para tiempos muy cercanos a $\tau$, hemos observado que la capacidad de integración númerica se trunca en $\tau_{[Ut]}\simeq$-0.798999 donde, si se consideran más cifras significativas de este valor, la solución de densidad comienza a aumentar su orden de magnitud hasta ~1e204$_{[U\rho]}$

Aún así, respondemos a la interrogante desde la intuición de la fenomenología: La presencia de una cantidad mayor o menor en la densidad tiene una proporcionalidad inversa con la edad del Universo al considerar que las densidades más altas se encuentran en etapas tempranas de su evolución y los estados más frios de baja densidad se perciben en etapas muy avanzadas.

#**Exercise 4**

Definimos la función de integración `friedmann_lemaitre` considerando la reducción de orden para la ecuación de aceleración.

$$\dot{a}=\frac{da}{dt}$$

$$\ddot{a}=\frac{d\dot{a}}{dt}=\frac{4\pi}{3}G(2\rho_\Lambda-\rho_m)a$$

```python
# @markdown Friedmann - Lemaitre Equations

def friedmann_lemaitre(t, y):

  a, da = y

  rho_m = rho_0m*(a**-3)

  da = da
  d2a = 4*np.pi*G*a*(2*rho_lambda - rho_m)/3

  return np.array([da, d2a])
```

Hacemos uso del método de integración **LSODA** del módulo `scipy.integration.solve_ivp` para determinar el valor $\tau$ de truncamiento númerico en el que, si se consideran mayores cifras significativas, la solución de aceleración adquiere valores negativos y su orden de magnitud diverge.

- **Universo Abierto:** Encontramos valor de truncamiento númerico $\tau_{[Ut]}\simeq -$0.805953 igual a ~11.26 Gyr

- **Universo Crítico:** Encontramos valor de truncamiento númerico $\tau_{[Ut]}\simeq -$0.959937 igual a ~13.41 Gyr. Este último valor se aproxima de mejor manera a la edad actual del Universo, indicando que la presencia de Energía Oscura se hace indispensable para comprender la evolución del Universo según las observaciones





```python
# @markdown Open Universe Solution

#Initial Conditions
Omega_0, Omega_0lambda = (0.3, 0)

rho_0c = 3*(H0**2)/(8*np.pi*G)

rho_lambda = Omega_0lambda*rho_0c

rho_0m = (Omega_0 - Omega_0lambda)*rho_0c

#Friedmann - Lemaitre Solution
universe_open = solve_ivp(friedmann_lemaitre, (0, -0.805953),
                          [a0 := 1, da0 := H0],
                          method = 'LSODA')

#Solution Values
ts_open = universe_open.t*Ut.to(u.Gyr) #Gyr
as_open, das_open = universe_open.y

#Mass Density and Hubble's Parameter
rhos_m_open = rho_0m*(as_open**-3)
Hs_open = das_open/as_open

#Cosmic Acceleration Gyr^-2
d2as_open = 4*np.pi*G*as_open*(2*np.full_like(as_open, rho_lambda) -\
                               rhos_m_open)/3/(Ut.to(u.Gyr)**2)

#Universe Age
print(f'Predicted Age for Universe:\n\nτ = {-ts_open[-1]} Gyr')
```

```python
# @markdown Critic Universe Solution

#Initial Conditions
Omega_0, Omega_0lambda = (1, 0.7)

rho_lambda = Omega_0lambda*rho_0c

rho_0m = (Omega_0 - Omega_0lambda)*rho_0c

#Friedman - Lemaitre Solution
universe_critic = solve_ivp(friedmann_lemaitre, (0, -0.959937),
                            [a0 := 1, da0 := H0], method = 'LSODA')

#Solution Values
ts_critic = universe_critic.t*Ut.to(u.Gyr) #Gyr
as_critic, das_critic = universe_critic.y

#Mass Density and Hubble's Parameter
rhos_m_critic = rho_0m*(as_critic**-3)
Hs_critic = das_critic/as_critic

#Cosmic Acceleration Gyr^-2
d2as_critic = 4*np.pi*G*as_critic*(2*np.full_like(as_critic,
                                                  rho_lambda) -\
                                  rhos_m_critic)/3/(Ut.to(u.Gyr)**2)

#Universe Age
print(f'Predicted Age for Universe:\n\nτ = {-ts_critic[-1]} Gyr')
```

Dada la naturaleza de expansión descelerada del Universo en esta solución que solo considera la presencia de materia bariónica, las soluciones en la aceleración adquieren valores negativos y por tanto mostramos dicha cantidad haciendo enfasis en que su solución es negativa. Esto con el fin de lograr su representación en escala logarítmica.

#**Exercise 5**

Sea la definición temporalmente dependiente del parámetro de Hubble

$$H(t)=\frac{\dot{a}}{a}=\frac{1}{a}\frac{da}{dt}$$

A partir del Teorema de la Función Inversa, es posible encontrar una relación para el parámetro de Hubble que dependa del Factor de Escala, es decir, una relación en la que a su vez el tiempo es dependiente de este factor.

$$a(t)=f(t)\quad\to\quad t(a)=f^{-1}(a)$$

Así mismo, se satisface que

$$\frac{dt}{da}=\left(\frac{da}{dt}\right)^{-1}$$

Bajo esta situación entonces $H(t)\to H(a)$ y por tanto se obtiene que

$$dt=\frac{da}{a\,H(a)}$$

El tiempo de Loopback Time se define entonces como el intervalo temporal que ha transcurrido, desde cierto valor de $a$ en el pasado, hacia el presente.

$$t_{LB}(a)=\Delta\,t=\int_{t(a)}^{t_0}dt'=t_0-t(a)$$

Por otra parte, esta integración se realiza igualmente desde el valor de $a$ en el pasado hacia el valor presente $a(t_0)=1$

$$t_{LB}(a)=t_0-t(a)=\int_a^1 \frac{da'}{a'\,H(a')}$$

#**Exercise 6**

Despejando $H$ de la ecuación de Friedmann-Lemaitre, e introduciendo el término de densidad de curvatura $\rho_k$

$$\begin{align*}
H^2&=\frac{8\pi G}{3}\left(\frac{3\epsilon}{4\pi Ga^2}+\rho_m+\rho_\Lambda\right)\\&=\frac{8\pi G}{3}\left(\rho_k+\rho_m+\rho_\Lambda\right)
\end{align*}\qquad\rho_k = \frac{3\epsilon}{4\pi Ga^2}$$

Como una proporción que compara cierto valor del parámetro de Hubble $H$ respecto a su valor actual $H_0$, tenemos la siguiente expresión

$$\begin{align*}
\frac{H^2}{H_0^2}&=\frac{8\pi G}{3H_0^2}\left(\rho_k+\rho_m+\rho_\Lambda\right)\\&=\frac{\rho_k+\rho_m+\rho_\Lambda}{\rho_{c0}}
\end{align*}\qquad\rho_{c0} = \frac{3H_0^2}{8\pi G}$$

Cada uno de estas densidades sigue una relación con el Factor de Escala de la siguiente forma

$$\rho_m=\rho_{0m}\left(\frac{a_0}{a}\right)^3\qquad\rho_k=\rho_{0k}\left(\frac{a_0}{a}\right)^2\qquad\rho_\Lambda=\rho_{0\Lambda}$$

El parámetro $\Omega$, que define la relación entre la densidad actual del Universo $\rho$ y la densidad de un Universo que se expande asintoticamente con geometría plana $\rho_c$, nos permite realizar la siguiente reescritura de la expresión anterior.

$$\Omega = \frac{\rho}{\rho_c}\quad\to\quad\frac{H^2}{H_0^2}=\Omega_{0}\left(\frac{a_0}{a}\right)^3+\Omega_{0k}\left(\frac{a_0}{a}\right)^2+\Omega_\Lambda$$

Finalmente, es imediato que una forma funcional del parámetro de Hubble relaciona condiciones iniciales del Universo junto al Factor de Escala.

$$H(a) = H_0\sqrt{\Omega_{0m}(a_0/a)^3 + \Omega_{0k}(a_0/a)^2 + \Omega_\Lambda}$$

La definición del Redshift $z$ dentro de la Cosmología sigue una relación con el Factor de Escala $a$

Así, es inmediata la sustitución dentro de la expresión del parámetro de Hubble

$$H(z) = H_0\sqrt{\Omega_{0m}(1+z)^3 + \Omega_{0k}(1+z)^2 + \Omega_\Lambda}$$

Anteriormente mostramos que una expresión para el tiempo de Loopback Time es

$$t_{LB}(a)=\int_a^1 \frac{da'}{a'\,H(a')}$$

Resulta también inmediato reescribir la integral, y sus límites de integración bajo la sustitución del Redshift, teniendo en cuenta que $a_0=1$

$$a = \frac{1}{1+z}\quad\to\quad da = \frac{-1}{(1+z)^2}\,dz$$

$$a=1\to z=0\,;\quad a=a\to z = z$$

$$\int_a^1 \frac{da'}{a'\,H(a')}=\frac{1}{H_0}\int_0^z\frac{dz'}{(1+z')\sqrt{\Omega_{0m}(1+z')^3 + \Omega_{0k}(1+z')^2 + \Omega_\Lambda}}$$

Definiendo finalmente al Tiempo de Hubble como $t_H = H_0^{-1}$, el tiempo de Loopback Time se expresa con dependencia del Redshift como

$$t_{LB}(z)=t_H\int_0^z\frac{dz'}{(1+z')\sqrt{\Omega_{0m}(1+z')^3 + \Omega_{0k}(1+z')^2 + \Omega_\Lambda}}$$

Conociendo que en el inicio del Universo se tenía un valor de $a=0\to z=\infty$, calculamos la Edad del Universo mediante la integración del Loopback Time siguiendo la expresión

$$\tau=t_H\int_0^\infty\frac{dz'}{(1+z')\sqrt{\Omega_{0m}(1+z')^3 + \Omega_{0k}(1+z')^2 + \Omega_\Lambda}}$$

Con esta solución obtenemos un valor de Edad del Universo igual a 13.46 Gyr.
