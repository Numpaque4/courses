En la siguiente celda realizamos la selección de los valores de Presión y Temperatura, junto con el ajuste de unidades.

```python
# @title

#Termodinamic Variables Profiles
rhos_emp = data["rho"]*1e3
Ts_emp = data["Talt"]
```

A continuación definimos el arreglo de la composición química atmosférica.

Realizamos el calculo de la Masa Molar.

```python
# @title

#Chemical Composition of Atmosphere
molecules = [dict(name = 'Ar', Mmol = element('Argon').atomic_weight),
            dict(name = 'N2', Mmol = 2*element('Nitrogen').atomic_weight),
            dict(name = 'O2', Mmol = 2*element('Oxygen').atomic_weight),
            dict(name = 'N', Mmol = element('Nitrogen').atomic_weight),
            dict(name = 'O', Mmol = element('Oxygen').atomic_weight),
            dict(name = 'H', Mmol = element('Hydrogen').atomic_weight),
            dict(name = 'He', Mmol = element('Helium').atomic_weight)]

#Molar Weight Calculation
Ms = np.zeros((len(molecules), N_zs, N_Lats, N_Longs))
ns = np.zeros((len(molecules), N_zs, N_Lats, N_Longs))

#Over all the Chemical Components
for i, mol in enumerate(molecules):
  ns[i] = data[mol['name']][0]
  Ms[i] = mol['Mmol']*ns[i]

#S.I Units
Ms_mol = Ms.sum(axis = 0)*1e-3/ns.sum(axis = 0)
```

De acuerdo con la ecuación de estado del Gas Ideal realizamos el cálculo de los perfiles de Presión

$$P=\frac{\rho RT}{M_{mol}}$$

```python
# @title

#Ideal Gas Pressure (Pa)
ps_emp = c.R.value*rhos_emp*Ts_emp/(Ms_mol)
```

#**Ejercicio 3**

Utilizando la siguiente expresión para gravedad variable con la altura

$$g(z)=\frac{g_0}{\left(1+\frac{z}{R_\oplus}\right)^2}$$

calculamos el valor del Gradiente de Presión ($-\rho(z)g(z)$) en cada una de las alturas para la fecha 4/7/2023 12:00:00 en la región del Mediterraneo.

Para poder realizar una evaluación de los valores de Presión empíricos, respecto a los valores de Presión que modela la ecuación de estado del Gas Ideal, planteamos la siguiente métodología para establecer el Gradiente de Presión.

- Tomando los valores de Presión calculados, establecemos una interpolación de modo que se tenga una función continua de Presión(`scipy.interpolate.interp1d`).
- Teniendo dicha función de interpolación, establecemos la función de diferenciación de Presión, respecto a la altura (`scipy.misc.derivative`)
-Ejecutamos la función de diferenciación sobre cada altura y realizamos el comparativo frente a los valores del Gradiente de Presión Empírico.

```python
# @title

#Gravity Variation
g = lambda z: c.g0.value/(1 + z/c.R_earth.value)**2

#Ideal Gas Pressure Variation
dps_ideal = np.array([derivative(interp1d(zs, ps_emp[0,:,12,30]), z)
                      for z in zs[1:-1]])

#Empirical Pressure Variation
dps_emp = (-rhos_emp[0,:,12,30]*g(zs))[1:-1]
```

En el siguiente Plot se observa como las curvas del Gradiente de Presión Numerico y Empírico se solapan sobre sí mismas. Esto nos indica que la condición de Equilibrio Hidrostatico se satisface para la Atmósfera en la Tierra bajo la suposición de una Atmósfera de Gas Ideal.

```python
# @title
fig = go.Figure(data = [go.Scatter(y = zs/1e3,
                                   x = dps_ideal/c.atm.value,
                                   name = 'Pressure Gradient'),
                        go.Scatter(y = zs/1e3,
                                   x = dps_emp/c.atm.value,
                                   name = 'Force Density',
                                   mode = 'lines')])

fig.update_layout(title = dict(text = 'Hidrostatic Equilibrium<br>'+\
                                      f'Ibiza, España<br>{str(day_hot)}',
                               x = 0.5, y = 0.96),
                  height = 600, width = 800,
                  yaxis_title = 'Height (km)',
                  xaxis_title = '(atm/km)')

fig.show()
```

#**Ejercicio 4**

Siguiendo la idea de definir una función continua y diferenciable en todo punto para la Temperatura, planteamos el calculo de la altura de la Tropopausa de la misma forma.

Con la función de diferenciación de Temperatura realizamos una evaluación de los puntos donde esta función toma valor cero (`scipy.optimize.root_scalar`) con el fin de determinar los puntos críticos donde la Temperatura tiene pendiente cero y cambio en la concavidad del perfil de Temperatura.

Esto lo realizamos para cada una de las regiones definidas del planeta y de la evaluación de estos puntos críticos tomaremos el primero de ellos bajo la noción de que corresponde al primer mínimo donde la Temperatura cambia su comportamiento decreciente a uno creciente. Este punto corresponderá al valor de altura de la Tropopausa

```python
# @title

#Raices de la Función Derivada de la Función Interpolación de Temperatura
zs_tropopause = np.array([[root_scalar(
                              lambda z: derivative(
                                  interp1d(zs/1e3,
                                  Ts_emp.loc[day_hot,:,lat,lon]),
                                  z),
                              bracket = [10, 30]).root
                           for lon in Longs] for lat in Lats])
```

#**Ejercicio 6**

A continuación definimos la función `homentropic(zs, h0,gamma, p0, rho0, T0)` que calcula los perfiles de Presión, Densidad y Temperatura bajo el siguiente modelo Homentropico de Atmósfera.

$$T(z)=T_0\left(1-\frac{z(\gamma -1)}{\gamma h_0}\right)$$

$$P(z)=P_0\left(1-\frac{z(\gamma -1)}{\gamma h_0}\right)^{\frac{\gamma}{\gamma-1}}$$

$$\rho(z)=\rho_0\left(1-\frac{z(\gamma -1)}{\gamma h_0}\right)^{\frac{1}{\gamma -1}}$$

```python
# @title

#Function to calculate the Homentropic Atmosphere
def homentropic(zs, h0, gamma, p0, rho0, T0):

  #Homentropic Profiles
  Ts_hom = T0*(1 - zs*(gamma-1)/(gamma*h0))
  ps_hom = p0*((1 - zs*(gamma-1)/(gamma*h0))**(gamma/(gamma-1)))
  rhos_hom = rho0*((1 - zs*(gamma-1)/(gamma*h0))**(1/(gamma-1)))

  return ps_hom, rhos_hom, Ts_hom
```

En esta celda definimos la función `chi_square(x, zs, ps, rhos, Ts)` la cual trae integrada la función `homentropic()` en su interior para determinar el estadístico $\chi^2$

- `x`: Tipo Array que contiene los valores de `(h0, gamma, T0)` dado que emplearemos esta función dentro de la rutina `scipy.optimize.minimize` para obtener el valor de `x` que minimiza  

$$
 \chi^2(h_0, \gamma, T_0) = \sum_i\frac{[T_\text{hom}(z_i)-T(z_i)]^2}{T(z_i)^2} + \sum_i\frac{[\rho_\text{hom}(z_i)-\rho(z_i)]^2}{\rho(z_i)^2}+ \sum_i\frac{[p_\text{hom}(z_i)-p(z_i)]^2}{p(z_i)^2}
 $$

```python
# @title

#Function to calculate the Chi Square
def chi_square(x, zs, ps, rhos, Ts):

  h0, gamma, T0 = x

  #Homentropic Profiles
  ps_hom, rhos_hom, Ts_hom = homentropic(zs, h0, gamma,
                                         ps[0], rhos[0], Ts[0])

  #Chi Square Calculation
  Ts_sum = np.sum(((Ts_hom - Ts)/Ts)**2)
  ps_sum = np.sum(((ps_hom - ps)/ps)**2)
  rhos_sum = np.sum(((rhos_hom - rhos)/rhos)**2)

  chi2 = Ts_sum + ps_sum + rhos_sum

  return chi2
```

En esta celda establecemos los valores de los parámetros libres
- $\gamma=1.4$ corresponde a valor del Indice Adiabático para el Aire Seco a 20° C
- La Escala Vertical $h_0$ es calculada para cada región definida del planeta tomando los valores empiricos iniciales de Presión y Densidad

  $$h_0=\frac{P_0}{g_0\rho_0}$$

- Los valores de Temperatura Superficial $T_0$ son tomados de los valores empiricos Atmosféricos

Usando la función `scipy.optimize.minimize` sobre todo el conjunto de Coordenadas, minimizamos los valores de `x = [h0, gamma, T0]` que mejor ajustan el modelo Hometrópico a los valores empíricos de la Atmósfera hasta la Troposfera.


```python
# @title
#Gamma Value Estimation (Dry Air at 20°C)
gamma_free = 1.40

#Vertical Scale estimation over all Coordinates
h0s_free = ps_emp[0,0,:,:]/(c.g0.value*rhos_emp[0,0,:,:])

#Indices para tomar solamente la Altura hasta la respectiva Tropopausa
index_zs = np.round(zs_tropopause*1e3//50).astype(int)

#Chi Square Minimization over all Coordinates
chis_min = np.array([[minimize(chi_square,

                               [float(h0s_free[i,j]), gamma_free,
                                float(Ts_emp[0,0,i,j])],

                               args = (zs[:index_zs[i,j]],
                                       ps_emp[0,:index_zs[i,j],i,j].to_numpy(),
                                       rhos_emp[0,:index_zs[i,j],i,j].to_numpy(),
                                       Ts_emp[0,:index_zs[i,j],i,j].to_numpy())).x

                     for j in range(N_Longs)] for i in range(N_Lats)])

#Selction of Parameters
h0s = chis_min[:,:,0]
gammas = chis_min[:,:,1]
T0s = chis_min[:,:,2]
```

Con la intención de verifical el correcto ajuste de los valores minimizados, calculamos los perfiles de Presión, Densidad y Temperatura del Modelo Homentropico para la fecha 4/7/2023 12:00:00 en la región del Mediterraneo

```python
# @title

#Homentropic Profiles
ps_hom, rhos_hom, Ts_hom = homentropic(zs[:index_zs[12,30]],
                                       h0s[12,30],
                                       gammas[12,30],
                                       float(ps_emp[0,0,12,30]),
                                       float(rhos_emp[0,0,12,30]),
                                       T0s[12,30])
```

A continuación analizamos el correcto ajuste de los Perfiles de Densidad, Presión y Temperatura construidos sobre los parametros minimizados.
