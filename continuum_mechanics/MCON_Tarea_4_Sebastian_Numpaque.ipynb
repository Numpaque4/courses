#**Importe de Librerias**

```python
# @title
import numpy as np
import pandas as pd

import astropy.constants as const

from scipy.integrate import odeint
from scipy.optimize import curve_fit
from scipy.optimize import root_scalar

import plotly.graph_objects as go
```



---

#**Datos PREM500**

```python
# @markdown Cargue de Datos
!wget http://ds.iris.edu/files/products/emc/data/PREM500/PREM500_IDV.csv
```

```python
# @markdown Visualización de Datos
data = pd.read_csv('PREM500_IDV.csv').iloc[1:].astype('float64'); data
```

Seleccionamos los valores de Radio $r$, Densidad $\rho$, Velocidad de las S-Waves $v_S$ y Velocidad de las P-Waves $v_P$.

**Nota:** Se toman las Velocidades Verticales para representar un Modelo 1D.

```python
# @title

#Radius m
rs_PREM = data['(index -> (radius'].to_numpy()/1e3

#Density kg/m^3
rhos_PREM = data['density'].to_numpy()

#S-Waves Velocity m/s
vs_S_PREM = data['Vsv'].to_numpy()

#P-Waves Velocity m/s
vs_P_PREM = data['Vpv'].to_numpy()
```



---

#**Exercise 1**

Calculamos los Modulos de Compresibilidad y Rigidez bajo las siguientes expresiones

$$K = \rho\left(v_P^2-\frac{4}{3}v_S^2\right)\qquad \mu=\rho v_S^2$$

```python
# @title

#Shear Modulus
mus_PREM = rhos_PREM*(vs_S_PREM**2)

#Bulk Modulus
Ks_PREM = rhos_PREM*(vs_P_PREM**2 - 4*(vs_S_PREM**2)/3)
```

Mediante `numpy.diff` calculamos la diferencia absoluta discreta entre los valores de densidad y tomando los 3 principales valores donde la diferencia es mayor. Estas amplias variaciones resultan en las regiones de diferenciación de las capas.

```python
# @title

#Index over Radius Array for Layers
layers = np.sort(np.argsort(abs(np.diff(rhos_PREM[:-10])))[-3:]) + 1

#Layers Radius
r_layers = rs_PREM[layers]

print(f'Inner Core Radius: {r_layers[0]} km')
print(f'Outter Core Radius: {r_layers[1]} km')
print(f'Inner Mantle Radius: {r_layers[2]} km')
```

**Nota:** La graficación de $K$ y $\mu$ se realiza en el siguiente literal para contrastar con el ajuste realizado.

#**Exercise 2**

Definimos la función `fit_modulus(rho, *args)` que modela los valores del Módulo de Compresibilidad y Rigidez bajo sus parámetros de ajuste.

- `args`: Tupla o Array que contiene los parámetros de ajuste ($a,\,n_K$) o ($b,\,n_\mu$) para $K$ y $\mu$, respectivamente

```python
# @markdown Modulus Curve Model
fit_modulus = lambda rho, c, n: c*(rho**n)
```

A través del modulo `scipy.optimize.curve_fit` realizamos el ajuste para los modelos del Modulo de Compresibilidad y Rigidez en las capas del Nucleo Externo y el Manto Interno

```python
# @markdown Bulk Modulus Parameters Fitting

#Outter Core
K_args_core = curve_fit(fit_modulus,
                        rhos_PREM[layers[0]:layers[1]],
                        Ks_PREM[layers[0]:layers[1]])[0]

#Inner Mantle
K_args_mantle = curve_fit(fit_modulus,
                          rhos_PREM[layers[1]:layers[2]],
                          Ks_PREM[layers[1]:layers[2]])[0]

#Evaluating Curve Fitting
K_fit = np.append(fit_modulus(rhos_PREM[layers[0]:layers[1]],
                              *K_args_core),
                  fit_modulus(rhos_PREM[layers[1]:layers[2]],
                              *K_args_mantle))

print('Bulk Modulus fit parameters (a, nK):\n' \
      f'\nInner Mantle: {K_args_mantle}\n' \
      f'\nOutter Core: {K_args_core}\n')
```

```python
# @markdown Shear Modulus Parameters Fitting

#Inner Mantle
mu_args_mantle = curve_fit(fit_modulus,
                           rhos_PREM[layers[1]:layers[2]],
                           mus_PREM[layers[1]:layers[2]])[0]

#Outter Core
mu_args_core = [0, 0]

#Evaluating Curve Fitting
mu_fit = np.append(fit_modulus(rhos_PREM[layers[0]:layers[1]],
                               *mu_args_core),
                   fit_modulus(rhos_PREM[layers[1]:layers[2]],
                               *mu_args_mantle))

print('Shear Modulus fit parameters (b, nμ):\n' \
      f'\nInner Mantle: {mu_args_mantle}\n' \
      f'\nOutter Core: {mu_args_core}\n')
```

```python
# @markdown Graficación de Modulos de Compresibilidad y Rigidez

fig = go.Figure(data = [go.Scatter(x = rs_PREM,
                                   y = Ks_PREM/1e9,
                                   name = 'K'),
                        go.Scatter(x = rs_PREM,
                                   y = mus_PREM/1e9,
                                   name = 'μ'),
                        go.Scatter(x = rs_PREM[layers[0]:layers[2]],
                                   y = K_fit/1e9,
                                   name = 'K Fit'),
                        go.Scatter(x = rs_PREM[layers[0]:layers[2]],
                                   y = mu_fit/1e9,
                                   name = 'μ Fit')])

fig.add_vrect(x0 = 0, x1 = r_layers[0],
              line_width = 0, fillcolor = 'red',
              opacity = 0.1, annotation_text = "Inner Core")

fig.add_vrect(x0 = r_layers[0], x1 = r_layers[1],
              line_width = 0, fillcolor = 'orange',
              opacity = 0.1, annotation_text = "Outter Core")

fig.add_vrect(x0 = r_layers[1], x1 = r_layers[2],
              line_width = 0, fillcolor = 'yellow',
              opacity = 0.1, annotation_text = "Inner Mantle")

fig.update_layout(width = 900, height = 600,
                  title_text = 'Bulk and Shear Modulus',
                  title_x = 0.5, title_y = 0.95,
                  xaxis_title = 'Radius [km]',
                  yaxis_title = '[GPa]',
                  legend = dict(orientation = 'h',
                                y = 1.08, x = 0.83))

fig.show()
```

#**Exercise 3 and 4**

Definimos la función `planetary_model(y, r, K_args)` que establece el sistema de ecuaciones diferenciales a solucionar.

- `K_args` : Parametros de ajuste de la curva del Módulo de Compresibilidad

Para considerar la diferenciación de capas, se añade el parámetro `K_args` con el fin de establecer sobre qué región se hará la integración.

```python
# @markdown ODE System
def planetary_model(y, r, K_args):

  p, M, rho = y

  g = const.G.value*M/(r**2)

  dp = -rho*g
  dM = 4*np.pi*r*r*rho
  drho = -rho*rho*g/fit_modulus(rho, *K_args)

  return np.array([dp, dM, drho])
```

Definimos la función `mars_interior(R, y0)` que realiza la solución de los perfiles para el interior de Marte considerando la diferenciación de capas.

- `R`: Radio del Planeta

- `y0`: Condiciones Iniciales en $p$, $M$ y $\rho$

```python
# @markdown ODE System Integration Function
def mars_interior(R, y0):

  #Mantle Integration Interval
  rs_mantle = np.linspace(R, 0.5*R, 1000)

  #Mantle ODE System Solution
  solution_mantle = odeint(planetary_model, y0,
                           rs_mantle, args = (K_args_mantle,))

  #Core-Mantle Initial Values
  y0_core = solution_mantle[-1,:]
  y0_core[-1] *= 1.7 #Density Increase

  #Core Integration Interval
  rs_core = np.linspace(0.5*R, 0.05*R, 1000)

  #Core ODE System Solution
  solution_core = odeint(planetary_model, y0_core,
                         rs_core, args = (K_args_core,))

  #Creating Solution Array
  rs_mars = np.append(rs_mantle, rs_core)
  solution_mars = np.vstack((solution_mantle, solution_core))

  return np.column_stack((rs_mars, solution_mars))
```

Aquí planteamos un método de determinación de raices para la función `mass_ratio` que toma la proporción entre la solución de Masa $M_r$ cercana al centro y la Masa $M_m$ del Planeta.

Teniendo presente que para cierto valor de $R$, la proporción $M_r/M_m$ toma un valor de cero, mediante el módulo `scipy.optimize.root_scalar` encontramos dicho valor de $R$

```python
# @markdown Finding Overshoot Radius

#Mars Initial Values
p0_mars = 0 #Pa
rho0_mars = 3000 #kg/m^3
M_mars = 0.107*const.M_earth.value
R_mars = 0.532*const.R_earth.value

y0 = [p0_mars, M_mars, rho0_mars]

#Finding Root for Mass Ratios
mass_ratio = lambda R: mars_interior(R, y0)[-1,2]/M_mars

R_overshoot = root_scalar(mass_ratio,
                          bracket = (0.96*R_mars, 1.04*R_mars)).root

print(f'Mars Overshooting Radius: {R_overshoot/R_mars} R_mars')
```

Finalmente generamos la solución de los perfiles de Masa, Presión, Densidad mediante `mars_interior` tomando $R$ como el valor del Radio de Overshooting encontrado.

Así mismo, calculamos los perfiles de Gravedad, Módulo de Compresibilidad y Rigidez, y Velocidad de Ondas Sismicas S y P.

**Nota:** Los perfiles de $K$ y $\mu$ son calculados bajo la función `fit_modulus` utilizando los valores de $\rho$ calculados para el interior de Marte y los parámetros de ajuste encontrados.

```python
# @markdown ODE System Solution and Mars Profiles
rs_mars, ps_mars, Ms_mars, rhos_mars = mars_interior(R_overshoot, y0).T

gs_mars = const.G.value*Ms_mars/rs_mars**2

Ks_mars = np.append(fit_modulus(rhos_mars[:999], *K_args_mantle),
                    fit_modulus(rhos_mars[999:], *K_args_core))

mus_mars = np.append(fit_modulus(rhos_mars[:999], *mu_args_mantle),
                     fit_modulus(rhos_mars[999:], *mu_args_core))

vs_S_mars = np.sqrt(mus_mars/rhos_mars)

vs_P_mars = np.sqrt(Ks_mars/rhos_mars + 4*vs_S_mars/3)
```

```python
# @markdown Plotting Mars Profiles

fig = go.Figure(data = [go.Scatter(x = rs_mars/1e3,
                                   y = ps_mars/1e10,
                                   name = 'p'),
                        go.Scatter(x = rs_mars/1e3,
                                   y = Ms_mars/1e23,
                                   name = 'M'),
                        go.Scatter(x = rs_mars/1e3,
                                   y = rhos_mars/1e3,
                                   name = 'ρ'),
                        go.Scatter(x = rs_mars/1e3,
                                   y = gs_mars,
                                   name = 'g')])

fig.add_vrect(x0 = 0, x1 = 0.5*R_overshoot/1e3,
              line_width = 0, fillcolor = 'red',
              opacity = 0.1, annotation_text = "Core")

fig.add_vrect(x0 = 0.5*R_overshoot/1e3, x1 = R_overshoot/1e3,
              line_width = 0, fillcolor = 'orange',
              opacity = 0.1, annotation_text = "Mantle")

fig.update_layout(width = 900, height = 600,
                  title_text = 'Pressure, Mass, Density ' \
                               'and Gravity Profiles',
                  title_x = 0.5, title_y = 0.95,
                  xaxis_title = 'Radius [km]',
                  yaxis_title = 'p [10 GPa],  M [1e23 kg],' \
                                '  ρ [1e3 kg/m^3],  g [m/s]',
                  legend = dict(orientation = 'h',
                                y = 1.08, x = 0.7))

fig.show()
```
