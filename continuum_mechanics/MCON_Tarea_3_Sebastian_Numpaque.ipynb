#**Exercise 1**

Comenzamos declarando las variables espaciales simbolicas `(X, Y, Z)` y el campo de desplazamientos `u_field` en esta celda

```python
# @title

#Variables Espaciales
X, Y, Z = sp.symbols('X Y Z', integer = True)

X_vec = sp.Matrix([X, Y, Z])

#Campo de Desplazamientos
u_field = sp.Matrix([X + Y*1e-3 + X*X*2e-4 - Z*2e-3,
                     1.001*Y - Y*Y*2e-4 + Z*4e-3,
                     X*X*3e-4 + Y*4e-4 + 0.998*Z]) - X_vec
```

Definimos los `np.array` de valores `(X_vals, Y_vals, Z_vals)` que adquieren cada una de las variables espaciales `(X, Y, Z)` y la malla de coordenadas `np.meshgrid` de cada una de las variables `(X_mesh, Y_mesh, Z_mesh)`

```python
# @title

#Dimensiones del Bloque
X_long, Y_long, Z_long = (2, 1, 0.2)

#Valores de las Variables Espaciales
NXs, NYs, NZs = (20, 10, 20)

dX, dY, dZ = (X_long/NXs, Y_long/NYs, Z_long/NZs)

X_vals = np.arange(0, X_long + dX, dX)
Y_vals = np.arange(0, Y_long + dY, dY)
Z_vals = np.arange(0, Z_long + dZ, dZ)

#Malla de Coordenadas Espaciales
X_mesh, Y_mesh, Z_mesh = np.meshgrid(X_vals, Y_vals, Z_vals)
```

La funcin `plot_field_XYZ()` establece el graficador del Campo Vectorial de Desplazamientos. Dicha funcin recibe los siguientes argumentos:

- `Plano`: Define el Plano sobre el que se visualiza el Campo de Desplazamientos

- `i, j, k`: Indice sobre los valores de `(X_vals, Y_vals, Z_vals)` de la variables espaciales `(X, Y, Z)`

- `u_vals`: Valores del Campo de Desplazamientos `u_field` evaluado sobre la malla de coordenadas del bloque.

**Nota:** Los valores de los vectores de desplazamiento han sido escalados por un factor de 30 para visualizar apropiadamente el Campo de Desplazamiento

```python
# @title
def plot_field_XYZ(Plano, i, j, k, u_vals):

  #Condicionales de Graficacin de los Campos
  if Plano == 'X-Y':

    x, y = X_mesh[:,:,0], Y_mesh[:,:,0]
    u, v = u_vals[0,0,:,:,k], u_vals[1,0,:,:,k]
    text = f'<br>Z = {round(Z_vals[k],2)} m'

  if Plano == 'Y-Z':

    x, y = Y_mesh[:,0,:], Z_mesh[:,0,:]
    u, v = u_vals[1,0,:,i,:], u_vals[2,0,:,i,:]
    text = f'<br>X = {round(X_vals[i],2)} m'

  if Plano == 'X-Z':

    x, y = X_mesh[0,:,:], Z_mesh[0,:,:]
    u, v = u_vals[0,0,j,:,:], u_vals[2,0,j,:,:]
    text = f'<br>Y = {round(Y_vals[j],2)} m'

  #Graficacion del Campo de Desplazamientos
  fig = ff.create_quiver(x, y, 30*u, 30*v, #Escalamiento por factor 30
                         line_width = 1, scale = 1,
                         arrow_scale = 0.5)

  fig.update_layout(width = 900, height = 600,
                    xaxis_title = f'{Plano[0]} [m]',
                    yaxis_title = f'{Plano[-1]} [m]',
                    title_text = f'Deformations Field ({Plano} Plane)' +\
                                 text,
                    title_x = 0.5, title_y = 0.93)

  fig.show()
```

Se ejecuta la funcin `interactive()` sobre la funcin graficadora `plot_field_XYZ()` con el siguiente tipo de variables interactivas:

- `Plano`: De tipo `Dropdown` con seleccin sobre el Plano en el que se desea visualizar el Campo de Desplazamiento

- `i,j,k`: De tipo `IntSlider` para variar los valores de `(X, Y, Z)` dependiendo el Plano seleccionado

- `u_vals`: De tipo `fixed`. Corresponde a un array de dimensin `(NXs, NYs, NZs)` sobre la malla de coordenadas espaciales del bloque

La evaluacin del Campo de Desplazamientos sobre la malla de coordenadas se realiza mediante la funcin `sp.lambdify()`


```python
# @title

#Grafico Interactivo
interactive(plot_field_XYZ,
            Plano = Dropdown(options = ['X-Y', 'Y-Z', 'X-Z']),
            i = IntSlider(min = 0, max = NXs,
                          description = 'X : i',
                          continuous_update = False),
            j = IntSlider(min = 0, max = NYs,
                          description = 'Y : j',
                          continuous_update = False),
            k = IntSlider(min = 0, max = NZs,
                          description = 'Z : k',
                          continuous_update = False),
            u_vals = fixed(sp.lambdify(X_vec, u_field)(X_mesh,
                                                       Y_mesh,
                                                       Z_mesh)))
```

Se puede verificar con la anterior visualizacin que el Medio Continuo se encuentra en equilibrio rotacional donde no se inducen Fuerzas de Torcin y uncamente se tiene Fuerzas de Desplazamiento Axial.

#**Exercise 2**

Se calcula el Tensor de Deformaciones Infinitesimal `u_tensor` de acuerdo a la expresin

$$u_{ij}=\frac{1}{2}(\partial_iu_j+\partial_ju_i)$$

```python
# @title

#Tensor de Deformaciones Infinitesimal
u_tensor = sp.Matrix([[0.5*(u_field[j].diff(X_vec[i]) +\
                            u_field[i].diff(X_vec[j]))
                      for j in range(3)] for i in range(3)])

display(sp.Eq(sp.Symbol('u_ij'), u_tensor, evaluate = False))
```

Aqui definimos la funcin auxiliar `plot_box(X, Y, Z)` para visualizar el bloque en 3D:

- `X, Y, Z`: Dimensiones del Bloque

```python
# @title

#Graficador de Bloque
def plot_box(X, Y, Z):

  dots = [[0,X,X,0,0],
          [0,0,Y,Y,0]]

  fig = go.Figure(data = [go.Scatter3d(x = dots[0]*2,
                                       y = dots[1]*2,
                                       z = [0]*5 + [Z]*5,
                                       showlegend = False,
                                       mode = 'lines',
                                       marker_color = 'black')] +\

                          [go.Scatter3d(x = [dots[0][i]]*2,
                                        y = [dots[1][i]]*2,
                                        z = [0,Z],
                                        showlegend = False,
                                        mode = 'lines',
                                        marker_color = 'black')
                          for i in range(4)])

  fig.update_layout(width = 900, height = 600,
                    scene = dict(aspectmode="data"))

  return fig
```

La funcin `plot_main_directions()` genera la visualizacin de los Vectores de Deformacin Principal en algun punto del bloque:

- `i, j, k`: Indice sobre los valores de `(X_vals, Y_vals, Z_vals)` de la variables espaciales `(X, Y, Z)`

- `tensor`: Tensor sobre el que se visualizarn las Direcciones Principales

La funcin `plot_main_directions()` evalua internamente el Tensor de Deformaciones `u_tensor` en un punto `(X, Y, Z)` mediante la funcin `sp.lambdify()`.

La diagonalizacin del Tensor de Deformaciones `u_tensor` se da mediante el atributo `self.diagonalize()` que retorna una Matriz de EigenVectores, `eigenvecs`, y una Matriz Diagonal de EigenValores, `eigenvals`

Una vez conocidos los `eigenvecs` que definen los Vectores de Deformacin Principal, son escalados por los correspondientes `eigenvals` absolutos.

**Nota**: Los `eigenvecs` tambin son escalados por factor de 300 para una mejor visualizacin de los Vectores.

**Nota:** La funcin `plot_main_directions()` ser nuevamente utilizada en **Exercise 4** para visualizar las Direcciones Principales de Esfuerzo Normal

```python
# @title

#Graficador de las Direcciones Principales de Deformacin
def plot_main_deformations(i, j, k, tensor, factor):

  #Punto (X,Y,Z) de Visualizacin
  r_dot = np.array([X_vals[i], Y_vals[j], Z_vals[k]])

  #EigenValores y EigenVectores en el Punto
  eigenvecs, eigenvals = tensor.subs(list(zip(X_vec,
                                                r_dot))).diagonalize()

  #Escalamiento de los EigenVectores
  main_vecs = np.array(factor*sp.Abs(eigenvals)*eigenvecs,
                       dtype = np.float64)

  #Graficador del Bloque
  fig = plot_box(X_long, Y_long, Z_long)

  #Graficacin de Direcciones Principales de Deformacin
  for i in range(3):

    fig.add_trace(go.Scatter3d(x = [r_dot[0], main_vecs[i,0] + r_dot[0]],
                               y = [r_dot[1], main_vecs[i,1] + r_dot[1]],
                               z = [r_dot[2], main_vecs[i,2] + r_dot[2]],
                               showlegend = False,
                               mode = 'lines',
                               line_width = 3.5))

    fig.update_layout(title_text = 'Main Directions<br>' +\
                                   f'X = {round(X_vals[i], 2)}m, ' +\
                                   f'Y = {round(Y_vals[j], 2)}m, ' +\
                                   f'Z = {round(Z_vals[k], 2)}m',
                      title_x = 0.5, title_y = 0.94)

  fig.show()
```

Se ejecuta la funcin `interactive()` sobre la funcin graficadora `plot_main_directions()` con las siguientes variables interactivas:

- `i,j,k`: De tipo `IntSlider` para variar los valores de `(X, Y, Z)` definiendo el punto sobre el que se visualizan los Vectores de Deformacin Principal.

- `tensor`: De tipo `fixed` tomando al Tensor de Deformaciones `u_tensor`

```python
# @title

#Grafico Interactivo
interactive(plot_main_deformations,
            i = IntSlider(min = 0, max = NXs,
                          description = 'X : i',
                          continuous_update = False),
            j = IntSlider(min = 0, max = NYs,
                          description = 'Y : j',
                          continuous_update = False),
            k = IntSlider(min = 0, max = NZs,
                          description = 'Z : k',
                          continuous_update = False),
            tensor = fixed(u_tensor),
            factor = fixed(300))
```

En la visualizacin, podemos apreciar la ortogonalidad de los Vectores de Deformacin Principal en cada uno de los puntos `(X, Y, Z)`. En primera instancia, se puede identificar que dichas direcciones no varan a lo largo de todos los puntos del bloque. Numricamente, durante las pruebas se notaron ligeras variaciones en las componentes pero con ordenes despreciables.

La invarianza de las Direcciones Principales de Deformacin indica que el Medio se encuentra bajo un Campo de Deformaciones sin Fuerzas de Torcin y nicamente se tienen Fuerzas de Desplazamiento Axial.

#**Exercise 3**

Se establece el valor del Modulo de Young y el Coeficiente de Poisson para el Aluminio como

$$E=70\,GPa\qquad\nu=0.33$$

Igualmente se calculan los coeficientes de Lam ($\mu,\,\lambda$)

$$\mu=\frac{E}{2(\nu+1)}\qquad\lambda=\frac{E\nu}{(1-2\nu)(\nu+1)}$$

Definidos estos coeficientes, se calcula el Tensor de Esfuerzos `sigma_tensor` siguiendo la relacin

$$\sigma_{ij}= 2\mu u_{ij}+\lambda\delta_{ij}u_{kk}$$

Finalmente, se calcula el Campo Escalar de Presin `pressure` como

$$p=\frac{1}{3}\sigma_{ii}$$

- Modulo de Young y Coeficiente de Poisson del Aluminio: https://www.mit.edu/~6.777/matprops/aluminum.htm

```python
# @title

E = 70e9 #Modulo de Young
v = 0.33 #Coeficiente de Poisson

#Coeficientes de Lam
mu = 0.5*E/(1+v)
Lambda = E*v/((1-2*v)*(1+v))

#Traza del Tensor de Deformaciones
trace = sp.Trace(u_tensor).simplify()

#Tensor de Estrs
sigma_tensor = sp.Matrix([[2*mu*u_tensor[i,j] +\
                           Lambda*sp.KroneckerDelta(i,j)*trace
                          for j in range(3)] for i in range(3)])

#Campo Escalar de Presin
pressure = -sp.Trace(sigma_tensor).simplify()/3

display(sp.Eq(sp.Symbol('p'), pressure, evaluate = False))
print('')
display(sp.Eq(sp.Symbol('sigma_ij'), sigma_tensor, evaluate = False))
```

Se define la funcin `plot_pressure()` que genera la visualizacin del Campo Escalar de Presin sobre un plano definido de las coordenadas espaciales `(X, Y, Z)`. La funcin recibe las siguientes variables:

- `Plano`: Define el Plano sobre el que se visualiza el Campo de Presin

- `i, j, k`: Indice sobre los valores de `(X_vals, Y_vals, Z_vals)` de la variables espaciales `(X, Y, Z)`

- `pressure_vals`: Valores del Campo de Presin `pressure` evaluado sobre la malla de coordenadas del bloque.

```python
# @title

#Funcin de Graficacin del Campo de Presin
def plot_pressure(Plano, i, j, k, pressure_vals):

  #Condicionales de valores del Campo para cada Plano
  if Plano == 'X-Y':

    x, y = X_vals, Y_vals
    z = pressure_vals[:,:,k]
    text = f'{Plano} Plane, Z = {round(Z_vals[k],2)} m'

  if Plano == 'Y-Z':

    x, y = Y_vals, Z_vals
    z = pressure_vals[:,i,:].T
    text = f'{Plano} Plane, X = {round(X_vals[i],2)} m'

  if Plano == 'X-Z':

    x, y = X_vals, Z_vals
    z = pressure_vals[j,:,:].T
    text = f'{Plano} Plane, Y = {round(Y_vals[j],2)} m'

  #Graficacin del Campo Escalar de Presin
  fig = go.Figure(data = [go.Contour(z = z/1e6, #MPa
                                     x = x,
                                     y = y,
                                     line_smoothing = 0.1,
                                     contours_coloring = 'heatmap')])

  fig.update_layout(width = 900, height = 600,
                    title = dict(text = f'Pressure Map (MPa)<br>' + text,
                                 x = 0.5, y = 0.94),
                    xaxis_title = f'{Plano[0]} [m]',
                    yaxis_title = f'{Plano[-1]} [m]',)

  fig.show()
```

Se ejecuta la funcin `interactive()` sobre la funcin graficadora `plot_pressure()` con las siguientes variables interactivas:

- `Plano`: De tipo `Dropdown` con seleccin sobre el Plano en el que se desea visualizar el Campo de Presin

```python
# @title
interactive(plot_pressure,
            Plano = Dropdown(options = ['X-Y', 'Y-Z', 'X-Z']),
            i = IntSlider(min = 0, max = NXs,
                          description = 'X : i',
                          continuous_update = False),
            j = IntSlider(min = 0, max = NYs,
                          description = 'Y : j',
                          continuous_update = False),
            k = IntSlider(min = 0, max = NZs,
                          description = 'Z : k',
                          continuous_update = False),
            pressure_vals = fixed(sp.lambdify(X_vec, pressure)(X_mesh,
                                                               Y_mesh,
                                                               Z_mesh)))
```

Para determinar el punto de mayor Presin en el bloque, planteamos la siguiente solucin:

Mediante el mdulo `scipy.optimize.minimize` realizamos la minimizacin del **Negativo del Campo de Presin** dentro de la frontera definida por los limites espaciales del bloque bajo el argumento `bounds`.

Realizado esto, conocemos el punto de mayor presin que corresponde al **Negativo del Mnimo** localizado

```python
# @title

#Minimizacin del Negativo de Presin
max_pressure = minimize(sp.lambdify([tuple(X_vec)], -pressure),
                        (X_vals[NXs//2], Y_vals[NYs//2], Z_vals[NZs//2]),
                        bounds = ((0, X_long), (0, Y_long), (0, Z_long)))

print('El punto (X, Y, Z) y valor P de mayor Presin son:\n\n' +\
      f'(X, Y, Z) = {tuple(max_pressure.x)}\n\n' +\
      f'P = {-max_pressure.fun/1e6} MPa')
```

#**Exercise 4**

- **Sobre la Diagonalizacin:**

Teniendo presente que la diagonalizacin de una Matriz $\mathbf{A}$ es tal que puede ser escrita mediante una Matriz de EigenVectores $\mathbf{P}$ y una Matriz Diagonal de EigenValores $\mathbf{D}$

$$\mathbf{A}=\mathbf{P}\mathbf{D}\mathbf{P^{-1}}$$

En el caso en que $\mathbf{A}$ es una Matriz Simtrica $\mathbf{A}=\mathbf{A^T}$, su diagonalizacin viene dada como

$$\mathbf{A}=\mathbf{P}\mathbf{D}\mathbf{P^T};\quad \mathbf{PP^T=I}$$

- **Sobre el Tensor de Esfuerzos**

Sobre cada punto dentro de un Medio Continuo existen al menos 3 Planos Principales en los que no ocurren Esfuerzos de Corte $\tau_n$ y donde unicamente se presentan Esfuerzos Normales $\sigma_n$ asociados al Vector Normal $\mathbf{\hat{n}}$ que define cada Plano.

Para un Tensor de Esfuerzos $\sigma$ asociado a un punto dentro de un Medio Continuo, existen entonces Direcciones Principales de Esfuerzo Normal que corresponden a los Eigenvectores y EigenValores del Tensor en dicho punto.

Si el Medio Continuo se encuentra en equilibrio rotacional, el Tensor de Esfuerzos es Simtrico $\sigma=\sigma^T$. As, conocer las Direcciones y Valor de Esfuerzo Normal se reduce a un problema de diagonalizacin del Tensor de Esfuerzos.

- **Sobre la Rotacin:**

El Tensor de Esfuerzos $\sigma$ aquiere una forma diagonal cuando las Direcciones Principales de Esfuerzo $\hat{e}_{\sigma i}$ coinciden con los Ejes del Sistema Coordenado $\hat{e_i}$. Es decir, la Transformacin de Rotacin es tal que los Vectores Cannicos de la Base $\hat{e_i}$ se orientan en la direccin de los Vectores de Esfuerzo Principal $\hat{e}_{\sigma i}$.

Dicha transformacin lleva asociada una Matriz de Rotacin tal que

$$\hat{e}_i=\mathbf{R}\,\hat{e}_{\sigma i}$$

Dado que los vectores $\hat{e}_i$ corresponden a los elementos de la Base Cannica Euclidea, se puede identificar que las columnas de la Matriz de Rotacin $\mathbf{R}$ corresponden a los Vectores de Esfuerzo Principal $\hat{e}_{\sigma i}$

$$\mathbf{R}=
\begin{pmatrix}
\vdots&\vdots&\vdots\\
\hat{e}_{\sigma 1}&\hat{e}_{\sigma 2}&\hat{e}_{\sigma 3}\\
\vdots&\vdots&\vdots
\end{pmatrix};\qquad\mathbf{R}=\mathbf{P^T}$$

A esto se agrega que, al ser $\hat{e}_{\sigma i}$ Vectores Ortonormales, la Matriz de Rotacin corresponde a una Transformacin Ortogonal tal que $\mathbf{R^{-1}}=\mathbf{R^T}$

Si se aplica dicha Rotacin al Tensor de Esfuerzos siguiendo la Regla de Transformacin de Tensores, entonces el Tensor Diagonal que se obtiene resulta simplemente la Matriz Diagonal de EigenValores $\mathbf{D_\sigma}$

$$\sigma=\mathbf{PD_\sigma P^T}$$

$$\mathbf{R\sigma R^T=D_\sigma}$$

- **Sobre la Presin**

La Presin como medida de la Traza del Tensor de Esfuerzos, es una cantidad invariante ante Rotaciones. En trminos generales, la forma en como se distribuye el Campo de Presin no depende de la orientacin del Medio Continuo

$$p=-\frac{1}{3}\textrm{Tr}(\sigma)=-\frac{1}{3}\textrm{Tr}(\mathbf{D_\sigma})$$

En general se satisface la siguiente propiedad sobre la Traza

$$\textrm{Tr}(\mathbf{AB})=\textrm{Tr}(\mathbf{BA})$$

De este modo, se demuestra que la Presin es una cantidad invariante ante la Rotacin

$$\begin{align*}
\textrm{Tr}(\mathbf{R\sigma R^T})&=\textrm{Tr}(\mathbf{D_\sigma})\\
&=\textrm{Tr}(\mathbf{R^TR\sigma})\\
&=\textrm{Tr}(\mathbf{\sigma})
\end{align*}$$

$$\textrm{Tr}(\mathbf{\sigma})=\textrm{Tr}(\mathbf{D_\sigma})$$

Esto se expresa como justificacin a que al momento de visualizar el Campo Escalar de Presin, los resultados obtenidos sern exactamente iguales a los obtenidos en **Exercise 3**

En esta celda se define la funcin `pressure_rotation(X)` que calcula la Presin con base en la Traza de la Matriz Diagonal de EigenValores que se obtiene luego de realizar la Rotacin del Tensor de Esfuerzos.

- `X`: Vector de Coordenadadas `(X, Y, Z)` a evaluar

**Nota:** Podr percatarse que la funcin `pressure_rotation()` no posee el signo negativo de la traza. Esto se hace con el fin de ingresar el **Negativo del Campo de Presin** para minimizacin posterior mediante `scipy.optimize.minimize`. Al evaluar los valores de Presin se incluye el signo negativo al resultado de la funcin `pressure_rotation()`

```python
# @title

#Funcin de Presin en un Punto
pressure_rotation = lambda X: np.sum(sigma_tensor.subs(
                                  list(zip(X_vec, X))
                                  ).eigenvals(multiple = True))/3
```

Se calculan los Esfuerzos Normales como los elementos en la diagonal del Tensor de Esfuerzos. La variable `normal_stress` almacena en un vector estos elementos

$$\sigma_n=\hat{n}^T\cdot\sigma\cdot\hat{n}$$

As mismo, se calcula tambin los Vectores de Esfuerzo de Corte `shear_vecs` como una Matriz donde cada columna corresponde al Vector $\tau_n$

$$\tau_{ij}=\sigma_{ij}-\sigma_{ij}\delta_{ij}$$

```python
# @title

#Esfuerzos Normales
normal_stress = sigma_tensor.diagonal()

#Esfuerzos de Corte
shear_vecs = sigma_tensor - sp.diag(*normal_stress)

display(sp.Eq(sp.Symbol('sigma_n'), normal_stress.T, evaluate = False))
print('')
display(sp.Eq(sp.Symbol('tau_ij'), shear_vecs, evaluate = False))
```

Se define la funcin `plot_stresses` que genera la visualizacin del Campo Escalar de Esfuerzos Normales y el Campo Vectorial de Esfuerzos de Corte sobre las caras superiores (X = 2m, Y = 1m, Z = 0.2m) e inferiores (X = 0m, Y = 0m, Z = 0m) del bloque.

- `Face`: Define el Plano de la Cara Principal a visualizar.

- `normal_vals`: Valores del Campo de Esfuerzos Normales evaluado sobre la malla de coordenadas del bloque

- `shear_vals`: Valores del Campo de Esfuerzos de Corte evaluado sobre la malla de coordenadas del bloque



```python
# @title

#Graficacin de Campo de Esfuerzos Normales y de Corte
def plot_stresses(Face, normal_vals, shear_vals):

  #Condicionales de valores de los Campos para cada Cara
  if Face == 'X-Y':

    x, y, X, Y = X_vals, Y_vals, X_mesh[:,:,0], Y_mesh[:,:,0]
    z_up, z_inf = normal_vals[0,2,:,:,-1], -normal_vals[0,2,:,:,0]
    U_up, V_up = shear_vals[0,2,:,:,-1], shear_vals[1,2,:,:,-1]
    U_inf, V_inf = -shear_vals[0,2,:,:,0], -shear_vals[1,2,:,:,0]

  if Face == 'Y-Z':

    x, y, X, Y = Y_vals, Z_vals, Y_mesh[:,0,:], Z_mesh[:,0,:]
    z_up, z_inf = normal_vals[0,0,:,-1,:].T, -normal_vals[0,0,:,0,:].T
    U_up, V_up = shear_vals[1,0,:,-1,:], shear_vals[2,0,:,-1,:]
    U_inf, V_inf = -shear_vals[1,0,:,0,:], -shear_vals[2,0,:,0,:]

  if Face == 'X-Z':

    x, y, X, Y = X_vals, Z_vals, X_mesh[0,:,:], Z_mesh[0,:,:]
    z_up, z_inf = normal_vals[0,1,-1,:,:].T, -normal_vals[0,1,0,:,:].T
    U_up, V_up = shear_vals[0,1,-1,:,:], shear_vals[2,1,-1,:,:]
    U_inf, V_inf = -shear_vals[0,1,0,:,:], -shear_vals[2,1,0,:,:]

  #Visualizacin en 4 Plots
  fig = make_subplots(rows = 2, cols = 2,
                      vertical_spacing = 0.1, horizontal_spacing = 0.1,
                      subplot_titles=('Upper Normal Stress (MPa)',
                                      'Upper Shear Field (MPa)',
                                      'Lower Normal Stress (MPa)',
                                      'Lower Shear Field (MPa)'),
                      x_title = f'{Face[0]} [m]',
                      y_title = f'{Face[-1]} [m]')

  #Visualizacin del Campo de Esfuerzo Normal y de Corte (Superior)
  fig.add_trace(go.Contour(z = z_up/1e6,
                           x = x,
                           y = y,
                           line_smoothing = 0.1,
                           contours_coloring = 'heatmap'),
                row = 1, col = 1)

  quiver_shear = ff.create_quiver(X, Y, U_up/1e6, V_up/1e6,
                                 line_width = 0.1, scale = 0.0005,
                                 arrow_scale = 0.15)

  fig.add_trace(go.Scatter(x = quiver_shear.data[0]['x'],
                           y = quiver_shear.data[0]['y'],
                           showlegend = False),
                row = 1, col = 2)

  #Visualizacin del Campo de Esfuerzo Normal y de Corte (Inferior)
  fig.add_trace(fig.data[0].update(z = z_inf/1e6),
                row = 2, col = 1)

  quiver_shear = ff.create_quiver(X, Y, U_inf/1e6, V_inf/1e6,
                                 line_width = 0.1, scale = 0.0005,
                                 arrow_scale = 0.15)

  fig.add_trace(go.Scatter(x = quiver_shear.data[0]['x'],
                           y = quiver_shear.data[0]['y'],
                           showlegend = False),
                row = 2, col = 2)

  #Plot Style
  fig.update_layout(width = 1400, height = 1000,
                    title_text = f'{Face} Face',
                    title_x = 0.5, title_y = 0.95)

  fig.show()
```

Se ejecuta la funcin `interactive()` sobre la funcin graficadora `plot_stresses()` con las siguientes variables interactivas:

- `Face`: De tipo `Dropdown` con seleccin sobre el Plano Principal de la Cara en el que se desean visualizar los Campos de Esfuerzo Normal y de Corte

- `normal_vals`: De tipo `fixed`. Corresponde a un array de dimensin `(NXs, NYs, NZs)` evaluando las componentes de Esfuerzo Normal `normal_stress` sobre la malla de coordenadas del bloque.

- `normal_vals`: De tipo `fixed`. Corresponde a un array de dimensin `(NXs, NYs, NZs)` evaluando los vectores de Esfuerzo de Corte `shear_vecs` sobre la malla de coordenadas del bloque.

La evaluacin de `normal_stress` y `shear_vecs` se realiza mediante `sp.lambdify()`

**Nota:** Podr percatarse que la evaluacin del Campo Vectorial de Esfuerzo de Corte incluye una operacin en la que se suman y se restan los valores de la coordenada espacial `X`.

Esto se realiza con el fin de que aquellas entradas en `shear_vecs` que no poseen dependencia de las variables espaciales `(X, Y, Z)` mantengan la forma de dimensin `(NXs, NYs, NZs)` al ser evaluadas.

```python
# @title

#Grafico Interactivo
interactive(
    plot_stresses,
    Face = Dropdown(options = ['X-Y', 'Y-Z', 'X-Z']),
    normal_vals = fixed(sp.lambdify(X_vec, normal_stress)(X_mesh,
                                                          Y_mesh,
                                                          Z_mesh)),
    shear_vals = fixed(sp.lambdify(X_vec,
                                    shear_vecs + X*sp.ones(3,3))(X_mesh,
                                                                 Y_mesh,
                                                                 Z_mesh) -\
                        np.array([[X_mesh]*3]*3)))
```
