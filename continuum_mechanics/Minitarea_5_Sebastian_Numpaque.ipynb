```python
# @title
import numpy as np
import pandas as pd

import plotly.graph_objects as go
```

#**Datos PREM500**

```python
# @markdown Visualización de Datos
data = pd.read_csv('PREM500_IDV.csv').iloc[1:].astype('float64'); data
```

Seleccionamos los valores de Radio $r$, Densidad $\rho$, Velocidad de las S-Waves $v_S$ y Velocidad de las P-Waves $v_P$.

**Nota:** Se toman las Velocidades Verticales para representar un Modelo 1D.

```python
# @title
#Radio m
radius = data['(index -> (radius'].to_numpy()

#Densidad kg/m^3
rhos = data['density'].to_numpy()

#Velocidad S-Waves m/s
vs_S = data['Vsv'].to_numpy()

#Velocidad P-Waves m/s
vs_P = data['Vpv'].to_numpy()
```



---

#**Exercise 1**

Calculamos los Modulos de Compresibilidad y Rigidez bajo las siguientes expresiones

$$K = \rho\left(v_P^2-\frac{4}{3}v_S^2\right)\qquad \mu=\rho v_S^2$$

```python
# @title
#Modulo de Cizalladura
mu = rhos*(vs_S**2)

#Modulo de Compresibilidad
K = rhos*(vs_P**2 - 4*(vs_S**2)/3)
```

```python
# @markdown Graficación de Modulos de Compresibilidad y Rigidez
fig = go.Figure(data = [go.Scatter(x = radius/1e3,
                                   y = K/1e6,
                                   name = '$K$'),
                        go.Scatter(x = radius/1e3,
                                   y = mu/1e6,
                                   name = '$\mu$')])

fig.update_layout(width = 900, height = 600,
                  title_text = 'Modulo de Compresibilidad y Rigidez',
                  title_x = 0.5, title_y = 0.95,
                  xaxis_title = 'Radius [km]',
                  yaxis_title = '[MPa]',
                  legend = dict(orientation = 'h',
                                y = 1.08, x = 0.83))

fig.show()
```



---

#**Exercise 2**

Calculamos el Modulo de Young, Coeficiente de Poisson y Coeficientes de Lamé bajo las siguientes expresiones

$$\nu = \frac{3K-2\mu}{6K+2\mu}\qquad E=2\mu(1+\nu)\qquad\lambda=\frac{E\nu}{(1-2\nu)(1+\nu)}$$

```python
# @title
#Coeficiente de Poisson
nu = (3*K - 2*mu)/(6*K + 2*mu)

#Modulo de Young
E = 2*mu*(1 +nu)

#Coeficiente de Lamé
lamda = E*nu/((1 - 2*nu)*(1 + nu))
```

```python
# @markdown Visualización de Valores Obtenidos
pd.DataFrame(list(zip(radius, K, E, nu, mu, lamda)),
             columns = ['R [m]', 'K [MPa]', 'E[MPa]',
                        'ν', 'μ [MPa]', 'λ [MPa]'])
```
